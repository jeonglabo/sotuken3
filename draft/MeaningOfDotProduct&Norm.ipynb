```python
import matplotlib.pyplot as plt
import numpy as np

# ベクトルの定義
a = np.array([1, 3])
b = np.array([5, 1])

# ベクトル a を b に射影する
projection_ab = (np.dot(a, b) / np.dot(b, b)) * b


# グラフの設定
plt.figure()

# ベクトル a, b, 射影ベクトルを描画
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector b')
plt.quiver(0, 0, projection_ab[0], projection_ab[1], angles='xy', scale_units='xy', scale=1, color='g', label='Projection of a onto b')

# 射影ベクトルからベクトルaの先端までの垂線を描画
plt.plot([projection_ab[0], a[0]], [projection_ab[1], a[1]], 'k--', label='Perpendicular from projection to a')

# 軸の設定
plt.xlim(0, 7)
plt.ylim(0, 4)
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# ラベルと凡例の設定
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Projection of Vector a onto Vector b with Perpendicular Line')

# グラフを表示
plt.show()
```

### ベクトルの内積と正射影

- ベクトル $\mathbf{a}$ を $\mathbf{b}$ に射影したものが $\mathbf{b} \hat{x}$ です。
- $\mathbf{a}$ と $\mathbf{b} \hat{x}$ は垂直であることを利用します。

$$
(\mathbf{a} - \mathbf{b} \hat{x}) \perp \mathbf{b} \hat{x}
$$

$$
\Rightarrow (\mathbf{a} - \mathbf{b} \hat{x})^T \mathbf{b} \hat{x} = 0
$$

ただし $\hat{x} \neq 0$ の場合、

$$
\mathbf{a}^T \mathbf{b} - \hat{x} \mathbf{b}^T \mathbf{b} = 0
$$

$$
\Rightarrow \hat{x} = \frac{\mathbf{a}^T \mathbf{b}}{\mathbf{b}^T \mathbf{b}}
$$

### ノルム (norm)

- **$2$-norm**: $\ell_2$-norm

$$
\|\mathbf{a}\| = \sqrt{\mathbf{a} \cdot \mathbf{a}} = \sqrt{x^2 + y^2 + z^2} = (x^2 + y^2 + z^2)^{\frac{1}{2}}
$$



$$
\mathbf{a} = \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix} \Rightarrow \|\mathbf{a}\|_2 = \sqrt{1^2 + 2^2 + 3^2} = \sqrt{14}
$$

$$
\mathbf{b} - \mathbf{a} = \begin{bmatrix} 1 \\ 2 \end{bmatrix} - \begin{bmatrix} 2 \\ 1 \end{bmatrix} = \begin{bmatrix} -1 \\ 1 \end{bmatrix}
$$

$$
\|\mathbf{b} - \mathbf{a}\| = \sqrt{(1-2) ^2+ (2-1)^2 } = \sqrt{(-1) ^2+ 1^2 }
$$



- **$1$-norm**: $\ell_1$-norm

$$
\|\mathbf{a}\| = |x| + |y| + |z|
$$



$$
\|\mathbf{a}\|_1 = 1 + 2 + 3 = 6
$$

- **$p$-norm**

$$
\left( |　|^p + |　|^p + |　|^p + \ldots \right) ^\frac{1}{p}
$$

$$
\|\mathbf{a}\|_p \approx \left( \sum |a_i|^p \right)^{\frac{1}{p}} \quad (p \ge 1)
$$

$$
p \to \infty
$$

$$
(|　|^\infty + |　|^\infty + |　|^\infty +  \ldots)^{\frac{1}{\infty}}
= 1 \quad \text{最も大きい値のみが残る}
$$

- **Infinity norm**

$$
\|\mathbf{a}\|_\infty = \max |a_i|
$$
