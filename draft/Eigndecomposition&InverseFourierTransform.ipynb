## 固有値分解と信号の変換

固有値分解は信号やベクトルの変換において重要な役割を果たします。特に、対称行列の固有値分解は直交行列による対角化が可能です。

### 固有値分解の表現

行列 $\mathbf{A}$ の固有値分解を次のように考えます：
$ \mathbf{A} = \lambda_1 \mathbf{q}_1 \mathbf{q}_1^T + \lambda_2 \mathbf{q}_2 \mathbf{q}_2^T + \lambda_3 \mathbf{q}_3 \mathbf{q}_3^T + \ldots $

ベクトル $\mathbf{x}$ を $\mathbf{A}$ で変換する場合、
$ \mathbf{A}\mathbf{x} = \lambda_1 \mathbf{q}_1 (\mathbf{q}_1^T \mathbf{x}) + \lambda_2 \mathbf{q}_2 (\mathbf{q}_2^T \mathbf{x}) + \lambda_3 \mathbf{q}_3 (\mathbf{q}_3^T \mathbf{x}) + \ldots $

$\mathbf{x}$ を固有ベクトルの線形結合で表すことができます。
$ \mathbf{x} = \sum_k \mathbf{q}_k (\mathbf{q}_k^T \mathbf{x}) $

---

### 具体例

例えば、
$ \mathbf{x} =
\begin{bmatrix}
-1 \\
2 \\
3
\end{bmatrix}
= 1
\begin{bmatrix}
1 \\
0 \\
0
\end{bmatrix}
+ 2
\begin{bmatrix}
0 \\
1 \\
0
\end{bmatrix}
+ 3
\begin{bmatrix}
0 \\
0 \\
1
\end{bmatrix} $

このベクトルの変換を $\mathbf{A}$ によって行うと、
$ \mathbf{A}\mathbf{x} = \mathbf{Q} \mathbf{\Lambda} \mathbf{Q}^T \mathbf{x} = \mathbf{x} $

---

### 信号の畳み込みと変換

連続信号 $x(t)$ とインパルス応答 $h(t)$ の畳み込みは次のように表されます：
$ (x * h)(t) = \int_{-\infty}^{\infty} x(\tau) h(t - \tau) d\tau $

これをフーリエ変換により表現すると、
$ X(\omega) H(\omega) = \int_{-\infty}^{\infty} x(t) e^{-j\omega t} dt \times \int_{-\infty}^{\infty} h(\tau) e^{-j\omega \tau} d\tau $

フーリエ変換の基底に基づき、この関係は信号の伝達関数を通じて処理されます。
$ y(t) = \frac{1}{T_0} \int_{-\infty}^{\infty} X(\omega) H(\omega) e^{j\omega t} d\omega $

---

### 信号の固有値分解

信号 $\mathbf{x}$ を固有ベクトルの成分で分解し、$\mathbf{A}$ による変換を通じて信号の変換を解析します。
$ \mathbf{x} = \sum_k \mathbf{q}_k \mathbf{q}_k^T \mathbf{x}, \quad \mathbf{A}\mathbf{x} = \sum_k \lambda_k \mathbf{q}_k (\mathbf{q}_k^T \mathbf{x}) $

この関係は信号処理やデータ圧縮において非常に重要で、信号の変換や特性解析に応用されます。
$ x(t) \to H(\omega) \to y(t) = \frac{1}{T_0} \int_{-\infty}^{\infty} X(\omega) H(\omega) e^{j\omega t} d\omega $

これにより、入力信号 $x(t)$ の特定の成分がどのように変換されて出力信号 $y(t)$ になるかが理解できます。
### 固有値分解 (Eigen Decomposition) と逆フーリエ変換

固有値分解は、逆フーリエ変換に対応する考え方です。行列 $\mathbf{A}$ の固有値分解は次のように表されます：
$ \mathbf{A} = \sum_k \lambda_k \mathbf{q}_k \mathbf{q}_k^T \implies \mathbf{A} \mathbf{x} = \sum_k \lambda_k \mathbf{q}_k (\mathbf{q}_k^T \mathbf{x}) $

ここで、$\mathbf{q}_k$ は固有ベクトル、$\lambda_k$ は固有値です。この分解は、システムの内部構造を示し、基底の内積に基づいてベクトルを展開します。

### 信号のフーリエ変換との関連

フーリエ変換を用いた信号処理では、入力信号 $h(t)$ は次のように表現されます：
$ h(t) = \int_{-\infty}^{\infty} H(\omega) e^{j\omega t} d\omega = \frac{1}{2\pi} \int_{-\infty}^{\infty} \langle h(t), e^{j\omega t} \rangle e^{j\omega t} d\omega $

この式は、固有値分解のアナロジーとして理解できます。

## 特異値分解 (SVD: Singular Value Decomposition)

特異値分解は、任意の行列 $\mathbf{A}$ を3つの行列の積に分解する手法です。以下の条件があるとき、特異値分解が可能です：

1. $\mathbf{A}$ が $m \times n$ の行列
2. $\mathbf{A}$ が対称行列の場合も含む
$ \mathbf{A} = \mathbf{U} \mathbf{\Sigma} \mathbf{V}^T $

ここで、$\mathbf{U}$ と $\mathbf{V}$ は直交行列で、$\mathbf{\Sigma}$ は対角行列です。対称行列の固有値分解と類似していますが、特異値分解は非対称行列にも適用できます。

- $\mathbf{A} = \mathbf{P} \mathbf{\Lambda} \mathbf{P}^{-1}$（対称行列の場合）
- $\mathbf{A} = \mathbf{Q} \mathbf{\Lambda} \mathbf{Q}^T$（対称行列の特別な場合）

### 直交行列の性質

$\mathbf{A}^T \mathbf{A}$ と $\mathbf{A} \mathbf{A}^T$ はいずれも対称行列です：
$ (\mathbf{A}^T \mathbf{A})^T = \mathbf{A}^T \mathbf{A}, \quad (\mathbf{A} \mathbf{A}^T)^T = \mathbf{A} \mathbf{A}^T $

特異値分解を行うと次のようになります：
$ \mathbf{A}^T \mathbf{A} = \mathbf{V} \mathbf{\Sigma} \mathbf{U}^T \mathbf{U} \mathbf{\Sigma} \mathbf{V}^T = \mathbf{V} \mathbf{\Sigma}^T \mathbf{\Sigma} \mathbf{V}^T $
