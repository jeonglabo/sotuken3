### 連立方程式の行列表現

連立方程式の例：

$$
\begin{cases}
  x + 2y = 4 \\
  -2x + 5y = 9
\end{cases}
\quad \Rightarrow \quad
\begin{bmatrix}
  1 & 2 \\
  2 & 5
\end{bmatrix}
\begin{bmatrix}
  x \\
  y
\end{bmatrix}
=
\begin{bmatrix}
  4 \\
  9
\end{bmatrix}
$$

もう一つの例：

$$
\begin{cases}
  x + 2y = 3 \\
  2x + 5y = 7
\end{cases}
\quad \Rightarrow \quad
\begin{bmatrix}
  1 & 2 \\
  2 & 5
\end{bmatrix}
\begin{bmatrix}
  x \\
  y
\end{bmatrix}
=
\begin{bmatrix}
  3 \\
  7
\end{bmatrix}
$$


これらの2つの連立方程式をまとめると
$$
\begin{bmatrix}
  1 & 2 \\
  2 & 5
\end{bmatrix}
\begin{bmatrix}
  x_1 & x_2 \\
  y_1 & y_2
\end{bmatrix}
=
\begin{bmatrix}
  4 & 3 \\
  2 & 7
\end{bmatrix}
$$

もっと増やすこともできる。
$$
\begin{bmatrix}
  1 & 2 \\
  2 & 5
\end{bmatrix}
\begin{bmatrix}
  x_1 & x_2 \\
  y_1 & y_2
\end{bmatrix}
=
\begin{bmatrix}
  4 & 3 & -2 & ... \\
  2 & 7 & 3 & ...
\end{bmatrix}
$$

行列の積に関する説明：

$$
m \times k \quad k \times n  \to m \times n
$$

このとき、掛ける順番に注意！

例：

$$
4 \times 3 \quad 3 \times 5 \quad 5 \times 2 \to 4 \times 2
$$

### 行列の性質

1. **内積として見る：$\mathbf{A}\mathbf{B}$**

$$
\mathbf{A} =
\begin{bmatrix}
  \mathbf{a}_1 \\
  \mathbf{a}_2 \\
  \mathbf{a}_3
\end{bmatrix}, \quad
\mathbf{B} =
\begin{bmatrix}
  \mathbf{b}_1 & \mathbf{b}_2 & \mathbf{b}_3
\end{bmatrix}
$$
<br>
$$
\mathbf{A}\mathbf{B} =
\begin{bmatrix}
  \mathbf{a}_1^T \cdot \mathbf{b}_1 & \mathbf{a}_1^T \cdot \mathbf{b}_2 & \mathbf{a}_1^T \cdot \mathbf{b}_3 \\
  \mathbf{a}_2^T \cdot \mathbf{b}_1 & \mathbf{a}_2^T \cdot \mathbf{b}_2 & \mathbf{a}_2^T \cdot \mathbf{b}_3 \\
  \mathbf{a}_3^T \cdot \mathbf{b}_1 & \mathbf{a}_3^T \cdot \mathbf{b}_2 & \mathbf{a}_3^T \cdot \mathbf{b}_3
\end{bmatrix}
$$

2. **Rank 1の行列の和**

$$
\mathbf{A}\mathbf{B} =
\begin{bmatrix}
  \mathbf{a}_1 & \mathbf{a}_2 & \mathbf{a}_3
\end{bmatrix}
\begin{bmatrix}
  \mathbf{b}_1 \\
  \mathbf{b}_2 \\
  \mathbf{b}_3
\end{bmatrix}
= \mathbf{a}_1 \mathbf{b}_1^T + \mathbf{a}_2 \mathbf{b}_2^T + \mathbf{a}_3 \mathbf{b}_3^T
$$

3. **Column Spaceとして見る**

$$
\mathbf{A}\mathbf{x} =
\begin{bmatrix}
  \mathbf{a}_1 & \mathbf{a}_2 & \mathbf{a}_3
\end{bmatrix}
\begin{bmatrix}
  x_1 \\
  x_2 \\
  x_3
\end{bmatrix}
= x_1 \mathbf{a}_1 + x_2 \mathbf{a}_2 + x_3 \mathbf{a}_3
$$
<br>
$$
\mathbf{A}\mathbf{B} =
\mathbf{A}
\begin{bmatrix}
  \mathbf{b}_1 & \mathbf{b}_2 & \mathbf{b}_3
\end{bmatrix} =
\begin{bmatrix}
  \mathbf{A}\mathbf{b}_1 & \mathbf{A}\mathbf{b}_2 & \mathbf{A}\mathbf{b}_3
\end{bmatrix}
$$

4. **Row Spaceとして見る**

$$
\mathbf{x}^T \mathbf{A} =
\begin{bmatrix}
  x_1 & x_2 & x_3
\end{bmatrix}
\begin{bmatrix}
  \mathbf{a}_1^T \\
  \mathbf{a}_2^T \\
  \mathbf{a}_3^T
\end{bmatrix} = x_1 \mathbf{a}_1 + x_2 \mathbf{a}_2 + x_3 \mathbf{a}_3
$$
