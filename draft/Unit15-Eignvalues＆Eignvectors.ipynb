### 固有値と固有ベクトル (Eigenvalue & Eigenvector)

固有値問題の定義：$\mathbf{A}\mathbf{x} = \lambda \mathbf{x}$

ここで、$\mathbf{A}$ は $n \times n$ の正方行列、$\mathbf{x}$ は $n \times 1$ の固有ベクトル、$\lambda$ は固有値です。

- $\mathbf{x} \to \mathbf{A} \mathbf{x}$ によって、線形変換が行われます。
- 固有ベクトル $\mathbf{x}$ は変換によって方向は変わらず、長さのみが $\lambda$ 倍されます。

#### 行列の例

行列 $\mathbf{A}$ について：

$ \mathbf{A} =
\begin{bmatrix}
2 & -1 \\
-1 & 2
\end{bmatrix} $

$\mathbf{A}$ を用いた変換：
- $\mathbf{A}\mathbf{x} =
\begin{bmatrix}
2 & -1 \\
-1 & 2
\end{bmatrix}
\begin{bmatrix}
2 \\
1
\end{bmatrix}
=
\begin{bmatrix}
3 \\
0
\end{bmatrix} $

- $\mathbf{A}\mathbf{x} =
\begin{bmatrix}
2 & -1 \\
-1 & 2
\end{bmatrix}
\begin{bmatrix}
1 \\
0
\end{bmatrix}
=
\begin{bmatrix}
2 \\
-1
\end{bmatrix} $

- $\mathbf{A}\mathbf{x} =
\begin{bmatrix}
2 & -1 \\
-1 & 2
\end{bmatrix}
\begin{bmatrix}
0 \\
1
\end{bmatrix}
=
\begin{bmatrix}
-1 \\
2
\end{bmatrix} $

これらの例では、変換されるベクトルはその向きと大きさが変化しています。この線形変換の一般的な例の中には、以下のような特別なケースが存在します。以下の例のように、線形変換によって、方向は変わらず、大きさのみが変化する場合があります。そのときのベクトルを固有ベクトルと呼び、変化率を固有値と呼びます。

- $\mathbf{A}\mathbf{x} =
\begin{bmatrix}
2 & -1 \\
-1 & 2
\end{bmatrix}
\begin{bmatrix}
1 \\
1
\end{bmatrix}
=
\begin{bmatrix}
1 \\
1
\end{bmatrix} \to \lambda = 1$

- $\mathbf{A}\mathbf{x} =
\begin{bmatrix}
2 & -1 \\
-1 & 2
\end{bmatrix}
\begin{bmatrix}
1 \\
-1
\end{bmatrix}
=
\begin{bmatrix}
3 \\
-3
\end{bmatrix} \
=
3\begin{bmatrix}
1 \\
-1
\end{bmatrix} \to \lambda = 3 $

これらの2つのベクトルを固有ベクトル(eignvectors)と呼び、一般に次のように表します。

$ \mathbf{x_1} =
\begin{bmatrix}
1 \\
1
\end{bmatrix} ,
\mathbf{x_2} =
\begin{bmatrix}
1 \\
-1
\end{bmatrix} $

線形変換行列$\mathbf{A}$は対称行列であることと、固有ベクトルが互いに直交していることに注目する必要があります。


### 行列の性質と線形変換

- 行列 $\mathbf{A}$ が固有ベクトル $\mathbf{x}_1, \mathbf{x}_2$ に対して同じ固有値 $\lambda$ を持つ場合、$\mathbf{x}_1$ と $\mathbf{x}_2$ が区別できなくなり、$\mathbf{A}$ は invertable ではない可能性があります。
$ \mathbf{A}
\begin{bmatrix}
1 & 1 \\
1 & 1
\end{bmatrix}
=
\begin{bmatrix}
2 \\
2
\end{bmatrix}
\quad (\text{invertible ではない}) $

---

### 固有関数と固有値の応用

連続信号 $x(t)$ と関数 $h(t)$ の畳み込みとしての固有関数の例：
$ x(t) \to h(t) \to x(t) * h(t) $

フーリエ変換に基づく表現： $ \int_{-\infty}^{\infty} h(\tau) e^{-j\omega_0 (\tau - t)} d\tau = e^{j\omega_0 t} \cdot \lambda $

ここで、$\lambda$ は固有値、$e^{j\omega_0 t}$ は固有関数として振る舞います。

---

固有値と固有ベクトルは、線形変換の方向性とスケーリングを理解するための重要なツールであり、物理学、エンジニアリング、データ解析など多くの分野で応用されています。
### 固有値と固有ベクトルの求め方

方程式 $\mathbf{A}\mathbf{x} = \lambda\mathbf{x}$ を整理すると、
$ (\mathbf{A} - \lambda \mathbf{I})\mathbf{x} = 0 $

となります。これは、固有値問題の一般形であり、$\mathbf{x} = 0$ は自明解 (trivial solution) です。非自明な解を求めるには、
$ \text{det}(\mathbf{A} - \lambda \mathbf{I}) = 0 $

を満たす $\lambda$ を見つける必要があります。

#### Null Spaceの理解

- $\mathbf{A} - \lambda \mathbf{I}$ の Null Space に属するすべてのベクトルが固有ベクトル $\mathbf{x}$ となります。
- この Null Space の基底 (basis) となるベクトルが固有ベクトルとして選ばれます。




### 例題 1
$ \mathbf{A} =
 \begin{bmatrix}
2 & -1 \\
-1 & 2
\end{bmatrix} $

固有値の計算：
$ \text{det}(\mathbf{A} - \lambda \mathbf{I}) =
\begin{vmatrix}
2 - \lambda & -1 \\
-1 & 2 - \lambda
\end{vmatrix} = (2 - \lambda)^2 - 1 = \lambda^2 - 4\lambda + 3 = 0 $

$ \lambda = 1, 3 $

それぞれの固有値に対応する固有ベクトル：

1. $ \lambda = 1$ の場合：$ (\mathbf{A} - \mathbf{I})\mathbf{x} =
   \begin{bmatrix}
1 & -1 \\
-1 & 1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} \implies \mathbf{x}_1 = \begin{bmatrix} 1 \\ 1 \end{bmatrix} $

2. $\lambda = 3$ の場合：

$ (\mathbf{A} - 3\mathbf{I})\mathbf{x} =
   \begin{bmatrix}
-1 & -1 \\
-1 & -1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} \implies \mathbf{x}_2 = \begin{bmatrix} 1 \\ -1 \end{bmatrix} $

線形変換行列
$ \mathbf{A} =
 \begin{bmatrix}
2 & -1 \\
-1 & 2
\end{bmatrix} $
を用いて任意のベクトル$\mathbf{x} =\begin{bmatrix}
2 \\
1
\end{bmatrix}$を線形変換することを考えます。このことに、この行列を成す、2つの列ベクトルを基底とする線形空間を考えます。そうすると、最初の列ベクトルの方向に2マスすすみ、2つ目の列ベクトルの方向に1マス進むと線形変換の結果の位置に到達することを意味します。すなわち、$x$は3で、$y$は0となります。

```python
import numpy as np
import matplotlib.pyplot as plt

# 定義された行列
A = np.array([[2, -1],
              [-1, 2]])

# グリッドの範囲
x_range = np.arange(-3, 5)
y_range = np.arange(-3, 5)

# 列ベクトル v1, v2 を取得
v1 = A[:, 0]
v2 = A[:, 1]

# グラフを描画
plt.figure(figsize=(8, 8))

# 基底ベクトル v1 を使ったグリッド
for x in x_range:
    plt.plot(v1[0]*x + v2[0]*y_range, v1[1]*x + v2[1]*y_range, 'r--', linewidth=1)
for y in y_range:
    plt.plot(v1[0]*x_range + v2[0]*y, v1[1]*x_range + v2[1]*y, 'r--', linewidth=1)

# 座標軸
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)

# 列ベクトルを描画
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='red', label='[3 0]')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='blue', label='[1 2]')

# ラベル
plt.xlabel('x')
plt.ylabel('y')

# グラフの範囲
plt.xlim(-3, 4)
plt.ylim(-3, 4)

# 表示
plt.grid(True)
plt.title("Grid in the Coordinate System Defined by Column Vectors of Matrix A")
plt.show()
```

固有値と固有ベクトルを計算して図示します。

```python
# 固有値と固有ベクトルを求める
eigenvalues, eigenvectors = np.linalg.eig(A)

# グラフのサイズ設定
plt.figure(figsize=(8, 8))

# 固有ベクトルを描画
for i in range(len(eigenvalues)):
    vec = eigenvectors[:, i] * 3  # 固有ベクトルを3倍して描画
    plt.quiver(0, 0, vec[0].real, vec[1].real, angles='xy', scale_units='xy', scale=1, label=f'Eigenvector {i+1}', color=['green', 'purple'][i])

# グラフの範囲
plt.xlim(-4, 8)
plt.ylim(-4, 8)

# 表示
plt.grid(True)
plt.title("Grid, Column Vectors of Matrix A, and Eigenvectors")
plt.show()

eigenvalues, eigenvectors
```

```python
# 固有値と固有ベクトルを求める
eigenvalues, eigenvectors = np.linalg.eig(A)

# グリッドの範囲
x_range = np.arange(-3, 5)
y_range = np.arange(-3, 5)

# 列ベクトル v1, v2 を取得
v1 = A[:, 0]
v2 = A[:, 1]

# グラフを描画
plt.figure(figsize=(8, 8))


# 基底ベクトル v1 を使ったグリッド
for x in x_range:
    plt.plot(v1[0]*x + v2[0]*y_range, v1[1]*x + v2[1]*y_range, 'r--', linewidth=1)
for y in y_range:
    plt.plot(v1[0]*x_range + v2[0]*y, v1[1]*x_range + v2[1]*y, 'r--', linewidth=1)

# 固有ベクトルを描画
for i in range(len(eigenvalues)):
    vec = eigenvectors[:, i] *eigenvalues[i]# 固有ベクトルを描画
    plt.quiver(0, 0, vec[0].real, vec[1].real, angles='xy', scale_units='xy', scale=1, label=f'Eigenvector {i+1}', color=['green', 'purple'][i])

# グラフの範囲
plt.xlim(-3, 5)
plt.ylim(-3, 5)

# 表示
plt.grid(True)
plt.title("Grid, Column Vectors of Matrix A, and Eigenvectors")
plt.show()

eigenvalues, eigenvectors
```

### 固有値分解 (Eigenvalue Decomposition)

2x2行列の固有値分解を考えます。行列 $\mathbf{A}$ が固有値と固有ベクトルを持つ場合：
$ \mathbf{A} \mathbf{x}_1 = \lambda_1 \mathbf{x}_1, \quad \mathbf{A} \mathbf{x}_2 = \lambda_2 \mathbf{x}_2 $

行列 $\mathbf{A}$ を次のように分解します：
$ \mathbf{A} [\mathbf{x}_1, \mathbf{x}_2] = [\mathbf{x}_1, \mathbf{x}_2]
\begin{bmatrix}
\lambda_1 & 0 \\
0 & \lambda_2
\end{bmatrix} $

$$ \mathbf{A} \mathbf{P} = \mathbf{P} \mathbf{\Lambda} $$

このとき、
$ \mathbf{P} = [\mathbf{x}_1, \mathbf{x}_2], \quad \mathbf{\Lambda} =
\begin{bmatrix}
\lambda_1 & 0 \\
0 & \lambda_2
\end{bmatrix} $

であり、固有値分解の式は以下のようになります。

$$ \mathbf{A} = \mathbf{P} \mathbf{\Lambda} \mathbf{P}^{-1} $$

$$ \mathbf{P}^{-1}\mathbf{A}\mathbf{P} =  \mathbf{\Lambda}  $$

これは、対角化可能な行列に適用され、固有ベクトルが線形独立である場合に使われます。固有値分解は行列の性質を解析する強力な手法です。

### 対角化
固有ベクトルが求まれば、$P$行列を求めることができ、その逆行列を求め、それらの行列を使用して対角化することができます。対角化することにより、線形変換行列によって固有ベクトルで形成される新しい基底では、どのような変換になるかが一目でわかるようになります。

```python
# 行列 A を対角化する
eigenvalues, eigenvectors = np.linalg.eig(A)

# 対角行列 D
D = np.diag(eigenvalues)

# 行列 P
P = eigenvectors

# 行列 P の逆行列
P_inv = np.linalg.inv(P)

D, P, P_inv

```

対角化

```python
D = P_inv @ A @ P

D
```


### 例題 2
$ \mathbf{A} =
\begin{bmatrix}
1 & 0 \\
0 & 2
\end{bmatrix} $

固有値の計算：
$ \text{det}(\mathbf{A} - \lambda \mathbf{I}) =
\begin{vmatrix}
1 - \lambda & 0 \\
0 & 2 - \lambda
\end{vmatrix} = (1 - \lambda)(2 - \lambda) = 0 $

$ \lambda = 1, 2 $

それぞれの固有値に対応する固有ベクトル：

1. $\lambda = 1$ の場合：

$ (\mathbf{A} - \mathbf{I})\mathbf{x} =
   \begin{bmatrix}
0 & 0 \\
0 & 1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} \implies \mathbf{x}_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix} $

2. $\lambda = 2$ の場合：
$ (\mathbf{A} - 2\mathbf{I})\mathbf{x} =
   \begin{bmatrix}
-1 & 0 \\
0 & 0
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} \implies \mathbf{x}_2 = \begin{bmatrix} 0 \\ 1 \end{bmatrix} $




### 例題 3
$ \mathbf{A} =
\begin{bmatrix}
2 & 0 \\
0 & 2
\end{bmatrix} $

固有値の計算：
$ \text{det}(\mathbf{A} - \lambda \mathbf{I}) =
\begin{vmatrix}
2 - \lambda & 0 \\
0 & 2 - \lambda
\end{vmatrix} = (2 - \lambda)^2 = 0 $

$ \lambda = 2 \quad (\text{重解}) $

対応する固有ベクトルは任意のスカラー倍でよく、例えば $\mathbf{x} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$ や $\mathbf{x} = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$ などが考えられます。





#### 例題 4
$ \mathbf{A} =
\begin{bmatrix}
0 & 1 \\
0 & 1
\end{bmatrix} $

固有値の計算：
$ \text{det}(\mathbf{A} - \lambda \mathbf{I}) =
\begin{vmatrix}
-\lambda & 1 \\
0 & 1 - \lambda
\end{vmatrix} = (1 - \lambda)(-\lambda) = 0 $

$ \lambda = 0, 1 $

それぞれの固有値に対応する固有ベクトル：

- $\lambda = 1$ の場合：

$ (\mathbf{A} - \mathbf{I})\mathbf{x} =
  \begin{bmatrix}
-1 & 1 \\
0 & 0
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} \implies \mathbf{x}_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix} $

この行列はシアー行列 (shearing matrix) であり、固有ベクトルの基底は1つだけです。

---

#### 例題 5

$ \mathbf{A} =
\begin{bmatrix}
0 & -2 \\
2 & 0
\end{bmatrix} $

固有値の計算：

$ \text{det}(\mathbf{A} - \lambda \mathbf{I}) =
\begin{vmatrix}
-\lambda & -2 \\
2 & -\lambda
\end{vmatrix} = \lambda^2 + 4 = 0 $

$ \lambda = \pm 2i \quad (\text{虚数解}) $

この行列は固有ベクトルを持たないため、ジョルダン標準形 (Jordan Canonical Form) などで扱います。





#### 例題 6

$ \mathbf{A} =
\begin{bmatrix}
1 & 1 \\
1 & 1
\end{bmatrix} $

固有値の計算：

$ \text{det}(\mathbf{A} - \lambda \mathbf{I}) =
\begin{vmatrix}
1 - \lambda & 1 \\
1 & 1 - \lambda
\end{vmatrix} = (1 - \lambda)^2 - 1 = \lambda^2 - 2\lambda = 0 $

$ \lambda = 0, 2 $

固有ベクトル：

- $\lambda = 0$ の場合：

$ (\mathbf{A} - 0\mathbf{I})\mathbf{x} =
  \begin{bmatrix}
1 & 1 \\
1 & 1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} \implies \mathbf{x}_1 = \begin{bmatrix} 1 \\ -1 \end{bmatrix} $

- $\lambda = 2$ の場合：

$ (\mathbf{A} - 2\mathbf{I})\mathbf{x} =
  \begin{bmatrix}
-1 & 1 \\
1 & -1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} \implies \mathbf{x}_2 = \begin{bmatrix} 1 \\ 1 \end{bmatrix} $

この行列はフルランクではありませんが、固有ベクトルを持つ特異なケースです。
