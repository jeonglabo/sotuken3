## ベクトル (Vector)

### 物理量の分類
- **スカラー (Scalar)**: 大きさのみ (例: 温度)
- **ベクトル (Vector)**: 大きさと方向 (例: 速度、力)

### ベクトルの表現
- ベクトル $\overrightarrow{v_1}$ は $\overrightarrow{AB}$ で表され、$\overrightarrow{AB} \neq \overrightarrow{BA}$。

### ベクトルの等価条件
- 大きさと方向が同じ $\rightarrow$ 同じベクトル

### ベクトルの演算
1. **加法**
   - 例: $\overrightarrow{v_1} + \overrightarrow{v_2}$
   - 図形的に矢印を平行移動して結ぶ。
   - 交換法則が成り立つ: $\overrightarrow{v_1} + \overrightarrow{v_2} = \overrightarrow{v_2} + \overrightarrow{v_1} = \overrightarrow{v_3}$

2. **引き算**
   - 例: $\overrightarrow{v_3} = \overrightarrow{v_1} - \overrightarrow{v_2}$
   - $\overrightarrow{v_3} = \overrightarrow{v_1} - \overrightarrow{v_2} = -(\overrightarrow{v_2} - \overrightarrow{v_1}) = \overrightarrow{v_1} + (-\overrightarrow{v_2})$

3. **スカラー倍**
   - 例: $2\overrightarrow{v_1}, \ \frac{1}{2}\overrightarrow{v_1}, -\overrightarrow{v_1}$
   - **スカラー倍 (Scalar Multiplication) **はベクトルの長さや向きを変える。

### 位置ベクトル
- 座標平面上の位置ベクトルは始点が原点 $(0, 0)$ である。
- 例: 点 $A(2, 2)$、点 $B(6, 4)$ の位置ベクトル $\overrightarrow{AB}$ は、$A$ から $B$ へのベクトル。

#### 座標変換

- $A$: $(x \rightarrow -2, y \rightarrow -2) \rightarrow (0, 0)$
- $B$: $(x \rightarrow -2, y \rightarrow -2) \rightarrow (4, 2)$

### 定理
- 点 $P_1 = (x_1, y_1)$、点 $P_2 = (x_2, y_2)$ の場合：
 $ \overrightarrow{P_1P_2} = (x_2 - x_1, y_2 - y_1) $  
## ベクトルの演算と性質

### ベクトルの演算
1. **加法**
   - ベクトル $\overrightarrow{v_1} = (a, b)$ と $\overrightarrow{v_2} = (c, d)$ の場合： $ \overrightarrow{v_1} + \overrightarrow{v_2} = (a + c, b + d) $

2. **引き算**
   - $\overrightarrow{v_1} - \overrightarrow{v_2} = (a_1, b_1) - (a_2, b_2) = (a_1 - a_2, b_1 - b_2)$

3. **スカラー倍**
   - $n\overrightarrow{v} = (na, nb)$

### ノルムと内積
1. **ノルム (ベクトルの長さ)**
   - ベクトル $\overrightarrow{v} = (v_1, v_2)$ の場合：$ \|\overrightarrow{v}\| = \sqrt{v_1^2 + v_2^2} $ - 一般のベクトル $\overrightarrow{v} = (v_1, v_2, \ldots, v_n)$ の場合： $ \|\overrightarrow{v}\| = \sqrt{\sum_{k=1}^{n} v_k^2} = \sqrt{v_1^2 + v_2^2 + \ldots + v_n^2} $

2. **ノルムの性質**
   - $\|\overrightarrow{v}\| \geq 0$
   - $\|\overrightarrow{v}\| = 0 \iff \overrightarrow{v} = \overrightarrow{0}$
   - $\|n\overrightarrow{v}\| = |n|\|\overrightarrow{v}\|$ （$n$はスカラー）

3. **単位ベクトル (Unit Vector)**
   - 長さが1のベクトル： $ \overrightarrow{u} = \frac{\overrightarrow{v}}{\|\overrightarrow{v}\|} $
   - 例：2次元では $(1, 0), (0, 1)$、3次元では $(1, 0, 0), (0, 1, 0), (0, 0, 1)$
   - 標準単位ベクトル：$\overrightarrow{e_1}, \overrightarrow{e_2}, \overrightarrow{e_3}$

### 距離 (Distance)
- 点 $A = (x_1, y_1)$ と点 $B = (x_2, y_2)$ の間の距離 $d$ は：
  $
  d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
  $

- ベクトルの始点と終点をそれぞれ $(x_1, y_1)$、$(x_2, y_2)$ としたとき：
  - $\overrightarrow{v_1} = (x_1, y_1)$
  - $\overrightarrow{v_2} = (x_2, y_2)$


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

from google.colab import drive
drive.mount('/content/drive')


# ベクトル場の設定
x, y = np.meshgrid(np.arange(0, 2.125, 0.125), np.arange(0, 2.125, 0.125))
u = 2 * x
v = np.zeros_like(y)

# ダイバージェンスの計算
div = np.gradient(u, axis=0) + np.gradient(v, axis=1)

# ベクトル場の描画
fig, ax = plt.subplots(figsize=(8, 8))
q = ax.quiver(x, y, u, v, color='b')
div_img = ax.imshow(div.T, extent=[0, 2, 0, 2], origin='lower', cmap='jet', alpha=0.6)
plt.colorbar(div_img, ax=ax, label='Divergence')
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')

# ベクトル場の更新用関数
def update_quiver(num, q):
    # ベクトルの大きさを変えてアニメーションを作成
    u = 2 * x * (1 + 0.1 * num)
    v = np.zeros_like(y)
    q.set_UVC(u, v)
    return q,

# アニメーションの作成と保存
ani = FuncAnimation(fig, update_quiver, fargs=(q,), frames=20, interval=200, blit=False)
ani.save('vector_field_animation.gif', writer='imagemagick', fps=10)
plt.show()

# xy平面で変化するベクトル場の設定
xx = np.linspace(-1, 10, 10)
yy = xx
x, y = np.meshgrid(xx, yy)
u = (x - 2) * (x - 8)
v = (y - 2) * (y - 8)

# ベクトル場の描画
fig2, ax2 = plt.subplots(figsize=(8, 8))
q2 = ax2.quiver(x, y, u, v)
div2 = np.gradient(u, axis=0) + np.gradient(v, axis=1)
div_img2 = ax2.imshow(div2.T, extent=[-1, 10, -1, 10], origin='lower', cmap='jet', alpha=0.6)
plt.colorbar(div_img2, ax=ax2, label='Divergence')
ax2.grid(True)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Vector Field with Divergence')

plt.show()

# ストリームラインのアニメーション
fig3, ax3 = plt.subplots(figsize=(8, 8))
stream = ax3.streamplot(x, y, u, v, color='k', density=1.5)
ax3.set_xlim([-1, 10])
ax3.set_ylim([-1, 10])
ax3.set_title('Streamlines with Moving Particles')

# パーティクルアニメーションの設定
particles, = ax3.plot([], [], 'ro', markersize=5)

def init_particles():
    particles.set_data([], [])
    return particles,

def update_particles(num):
    # パーティクルを更新してストリームラインに沿って移動させる
    x_particles = np.linspace(-1, 10, num)
    y_particles = np.sin(0.1 * num) * x_particles
    particles.set_data(x_particles, y_particles)
    return particles,

# ストリームパーティクルのアニメーション
ani_particles = FuncAnimation(fig3, update_particles, frames=100, init_func=init_particles, interval=50, blit=False)
ani_particles.save('/content/drive/MyDrive/Colab Notebooks/LinearAlgebra/streamline_particles_animation.gif', writer='imagemagick', fps=20)
plt.show()
```
