## コーシー・シュワルツの不等式 (Cauchy-Schwarz Inequality)

$
(\lambda x + \mu y)^2 \leq (\lambda^2 + \mu^2)(x^2 + y^2)
$

ベクトル $\mathbf{u} = (a, b)$ , $ \mathbf{v} = (x, y)$ に対して：
$
\|\mathbf{u}\|^2 \|\mathbf{v}\|^2
$

- 内積の表記：$ \mathbf{u} \cdot \mathbf{v} = ax + by $
- コーシー・シュワルツ不等式：
  $
  |\mathbf{u} \cdot \mathbf{v}| \leq \|\mathbf{u}\| \|\mathbf{v}\|
  $

- 証明：

  $
  \mathbf{u} \cdot \mathbf{v} = \|\mathbf{u}\| \|\mathbf{v}\| \cos \theta
  $
  
  ここで、$ -1 \leq \cos \theta \leq 1 $ より、両辺に $ \|\mathbf{u}\| \|\mathbf{v}\| $ をかけると、

  $
  -\|\mathbf{u}\| \|\mathbf{v}\| \leq \mathbf{u} \cdot \mathbf{v} \leq \|\mathbf{u}\| \|\mathbf{v}\|
  $

  したがって、

  $
  |\mathbf{u} \cdot \mathbf{v}| \leq \|\mathbf{u}\| \|\mathbf{v}\|
  $

## 三角不等式 (Triangle Inequality)

- 三角形の辺の長さの関係：
  $
  c \leq a + b
  $

- ベクトルの和に対する不等式：
  $
  \|\mathbf{u} + \mathbf{v}\|^2 \leq (\|\mathbf{u}\| + \|\mathbf{v}\|)^2
  $

- 証明：

  $
  (\mathbf{u} + \mathbf{v}) \cdot (\mathbf{u} + \mathbf{v}) = \mathbf{u} \cdot \mathbf{u} + 2\mathbf{u} \cdot \mathbf{v} + \mathbf{v} \cdot \mathbf{v} \leq \|\mathbf{u}\|^2 + 2\|\mathbf{u}\|\|\mathbf{v}\| + \|\mathbf{v}\|^2 = (\|\mathbf{u}\| + \|\mathbf{v}\|)^2
  $

- 三角不等式が成立するため、ベクトルの和の長さは各ベクトルの長さの和以下となります：

  $
  \|\mathbf{u} + \mathbf{v}\| \leq \|\mathbf{u}\| + \|\mathbf{v}\|
  $

## ピタゴラスの定理

直角三角形において：

$
\|\mathbf{u} + \mathbf{v}\|^2 = (\mathbf{u} + \mathbf{v}) \cdot (\mathbf{u} + \mathbf{v}) = \mathbf{u} \cdot \mathbf{u} + 2\mathbf{u} \cdot \mathbf{v} + \mathbf{v} \cdot \mathbf{v} = \|\mathbf{u}\|^2 + 0 + \|\mathbf{v}\|^2 = \|\mathbf{u}\|^2 + \|\mathbf{v}\|^2
$

ベクトル $ \mathbf{u} - \mathbf{v} = \|\mathbf{u}\|^2 + \|\mathbf{v}\|^2 $ となり、$ \cos 90^\circ = 0 $ の場合、この関係はピタゴラスの定理に従います。

## 直交条件（Orthogonality）

- **二つのベクトルが垂直のときの内積**
  
  直交するベクトル $ \mathbf{u} $ と $ \mathbf{v} $ の内積は 0 になる：
  
  $
  \mathbf{u} \cdot \mathbf{v} = 0
  $
  
  - 例: $ \mathbf{u} = 0 $ の場合は常に成り立つ。
  - 一般に $ \|\mathbf{u}\| \|\mathbf{v}\| \cos \theta = 0 $ で、$ \theta = 90^\circ $ のとき直交する。
