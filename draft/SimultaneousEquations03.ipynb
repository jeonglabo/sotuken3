このページでは、行列に対する基本的な行操作の具体的な例として、行の掛け算、行の入れ替え、行の加算について説明しています。

### 1. Row Multiplication (行の掛け算)
- 行列 $ E $ は行の掛け算を表現する行列です。
- 例えば、行列 $ EA $ は $ E $ を用いて元の行列 $ A $ の2行目を $ s $ 倍することを示しています。
- 行列 $ E $ の形は、単位行列の2行目の要素を $ s $ に変更したものです。
- 行列の積 $ EA $ によって、2行目の全ての要素が $ s $ 倍されます。

### 2. Row Switching (行の入れ替え)
- 行の入れ替え行列 $ P $ を用いて、特定の行を他の行と入れ替える操作が示されています。
- $ P_{32} $ は行列の3行目と2行目を入れ替える行列で、単位行列の対応する行を入れ替えることによって得られます。
- 入れ替え行列 $ P_{32} $ を行列 $ A $ にかけると、$ A $ の3行目と2行目が入れ替わります。

### 3. Row Addition Matrices (行の加算行列)
- 行の加算行列は、ある行に他の行の定数倍を加える操作を表現します。
- 例として、行列 $ EA $ を用いて、1行目に対して2行目を $ s $ 倍して加える操作が示されています。
- この操作は、行列 $ E $ の形で表され、$ s $ は特定の行のウェイト（加える倍率）として扱われます。
- 最終的に行列 $ EA $ は、各要素に対して行の加算が適用された新しい行列になります。

### まとめ
- 基本的な行列の操作（掛け算、入れ替え、加算）を行列形式で表すことで、複雑な計算を簡便に行うことができます。
- これらの操作は、行列の計算や連立方程式の解法など、多くの数学的応用において基本的な役割を果たします。


このページでは、行列操作を用いた連立方程式の解法について説明しています。以下がその内容の解説です。

### 行列の行操作と連立方程式の解法

#### 1. 行列 $ A $ と行操作
- 行列 $ A $ は以下のように定義されています：
  $$
  A =
  \begin{bmatrix}
  1 & 2 & 3 & 4 \\
  -1 & -2 & 2 \\
  0 & 3 & 2 & 1
  \end{bmatrix}
  $$

- $ E_A $ 行列を用いて $ A $ に対する行操作を行っています。例えば、行操作行列 $ E $ を $ A $ にかけることで、特定の行を調整します。

#### 2. 具体的な連立方程式の解法
- 例として、次の連立方程式が与えられています：
  $$
  \begin{cases}
  2x + 3y = 1 \\
  4x + y = 3
  \end{cases}
  $$

- これを行列形式で表すと：
  $$
  [A|b] =
  \begin{bmatrix}
  2 & 3 & 1 \\
  4 & 1 & 3
  \end{bmatrix}
  $$

- 行操作を行い、$ r_2 \to r_2 - 2r_1 $ を実行すると、次の行列が得られます：
  $$
  E_1 =
  \begin{bmatrix}
  1 & 0 \\
  -2 & 1
  \end{bmatrix}
  , \quad
  E_1[A|b] =
  \begin{bmatrix}
  2 & 3 & 1 \\
  0 & -5 & 1
  \end{bmatrix}
  $$

- 次に、$ r_1 \to r_1 - 3r_2 $ を行うと、行列は次のようになります：
  $$
  E_2 =
  \begin{bmatrix}
  1 & -3 \\
  0 & 1
  \end{bmatrix}
  , \quad
  E_2E_1[A|b] =
  \begin{bmatrix}
  2 & 0 & -2 \\
  0 & -5 & 1
  \end{bmatrix}
  $$

- 最後に、行のスケーリング操作を適用します：
  $$
  E_3 =
  \begin{bmatrix}
  \frac{1}{2} & 0 \\
  0 & \frac{1}{-5}
  \end{bmatrix}
  , \quad
  E_3E_2E_1[A|b] =
  \begin{bmatrix}
  1 & 0 & -1 \\
  0 & 1 & 1
  \end{bmatrix}
  $$

- これにより、解 $ x = -1, y = 1 $ が得られます。

### まとめ
この手順では、行列の行操作を駆使して連立方程式を解き、基本的な操作（行の加算、スケーリング、入れ替え）を行うことで簡単な形に変形し、解を得る方法を示しています。
