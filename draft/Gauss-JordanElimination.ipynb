
### ガウス・ジョルダン消去法 (Gauss-Jordan Elimination)

手順：

1. 両辺に定数をかける
2. 定数倍した行（列）から他の行（列）を引く／足す
3. 行（列）を入れ替える

例：

$$
\begin{bmatrix}
1 & 2 \\
2 & 5
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
4 \\
9
\end{bmatrix}
\quad \Rightarrow \quad
\text{拡張行列 (Extension Matrix)}
\quad
\begin{bmatrix}
1 & 2 & \vert & 4 \\
2 & 5 & \vert & 9
\end{bmatrix}
$$

ガウス・ジョルダン消去法により：

$$
\begin{bmatrix}
1 & 2 & 4 \\
2 & 5 & 9
\end{bmatrix}
\rightarrow
\begin{bmatrix}
1 & 2 & 4 \\
0 & 1 & 1
\end{bmatrix}
\rightarrow
\begin{bmatrix}
1 & 0 & 2 \\
0 & 1 & 1
\end{bmatrix}
$$

最終的な解は：

$$
x = 2, \quad y = 1
$$

### 例題：連立方程式の解法

方程式のセット：
$ \begin{cases}
2x + y - z = 8 \\
-3x - y + 2z = -11 \\
-2x + y + 2z = -3
\end{cases} $

行列の形で表す：
$ \begin{bmatrix}
2 & 1 & -1 & 8 \\
-3 & -1 & 2 & -11 \\
-2 & 1 & 2 & -3
\end{bmatrix} $

1行目を用いて行の操作を行います。
$ \begin{bmatrix}
2 & 1 & -1 & 8 \\
0 & 1/2 & 1/2 & 1 \\
0 & 2 & 1 & -3
\end{bmatrix} \quad (\text{行操作：} \text{①} + \text{②}) $
$ \begin{bmatrix}
2 & 1 & -1 & 8 \\
0 & 1/2 & 1/2 & 1 \\
0 & 0 & 1 & -1
\end{bmatrix} \quad (\text{行操作：} \text{③} - 2 \times \text{④}) $

最後に解を得る：
$ \begin{bmatrix}
1 & 0 & 0 & 2 \\
0 & 1 & 0 & 3 \\
0 & 0 & 1 & -1
\end{bmatrix} $

結果：
$ x = 2, \quad y = 3, \quad z = -1 $

### 逆行列 (Inverse Matrix)

$2 \times 2$ の行列の逆行列は以下のように求められます。
$ \begin{bmatrix}
a & b \\
c & d
\end{bmatrix}^{-1} = \frac{1}{ad - bc}
\begin{bmatrix}
d & -b \\
-c & a
\end{bmatrix} $

逆行列の求め方（行列を拡張してガウス・ジョルダン法を適用）：

1. 行列を拡張します。
   $ \begin{bmatrix}
   a & b & 1 & 0 \\
   c & d & 0 & 1
   \end{bmatrix} $

2. 行の操作を行い、左側を単位行列にします。
$ \rightarrow
   \begin{bmatrix}
   1 & 0 & \frac{d}{ad - bc} & \frac{-b}{ad - bc} \\
   0 & 1 & \frac{-c}{ad - bc} & \frac{a}{ad - bc}
   \end{bmatrix} $

最終的な逆行列：
$ \mathbf{A}^{-1} = \frac{1}{ad - bc}
\begin{bmatrix}
d & -b \\
-c & a
\end{bmatrix} $

ここで、$ad - bc$ は行列の行列式（determinant）を表し、$ad - bc = 0$ の場合は逆行列が存在しないことを意味します。

### 正方行列 $\mathbf{A}$ が可逆 (invertible) である条件

- **Non-singular Matrix**：行列が可逆の場合
- **Singular Matrix**：行列が可逆でない場合

可逆行列の条件：

1. $\text{det}(\mathbf{A}) \neq 0$
2. $\mathbf{A}$ が Full Rank である（すなわち、$\text{det}(\mathbf{A}) = 0$ の場合は $\mathbf{A}$ は Rank-deficient で Full Rank ではない）
3. Null Space ($\mathcal{N}(\mathbf{A})$) がゼロベクトルのみを含む

---

### 逆行列の性質

1. $(\mathbf{AB})^{-1} = \mathbf{B}^{-1} \mathbf{A}^{-1}$

   - $\mathbf{C} = \mathbf{AB}$ の場合、逆行列 $\mathbf{C}^{-1}$ が存在すれば、$\mathbf{CC}^{-1} = \mathbf{C}^{-1}\mathbf{C} = \mathbf{I}$。

2. $(\mathbf{A}^{-1})^{-1} = \mathbf{A}$

3. $(k\mathbf{A})^{-1} = \frac{1}{k} \mathbf{A}^{-1}$

   - ただし、$k$ はスカラー。

4. $(\mathbf{A}^T)^{-1} = (\mathbf{A}^{-1})^T$

5. $\text{det}(\mathbf{A}^{-1}) = \frac{1}{\text{det}(\mathbf{A})}$

---

### 行列式 (Determinant)

3x3行列 $\mathbf{A}$ の行列式の計算：
$ \mathbf{A} =
\begin{bmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{bmatrix} $
<br>
$ \text{det}(\mathbf{A}) = a(ei - fh) - b(di - fg) + c(dh - eg) $

- 余因子 (cofactor) を用いた計算。
