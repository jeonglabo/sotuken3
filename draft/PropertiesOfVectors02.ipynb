### 1. ベクトルの基本演算
- **ベクトル (Vector)**: 大きさ（Magnitude）と向き（Phase）を持つ量。座標系の変換に対して不変です。
- 図中では、ベクトル $\mathbf{v} = 3\mathbf{i} + 4\mathbf{j}$ が示されています。これは $x$ 軸と $y$ 軸に沿った成分を持つベクトルの表現です。

### 2. 数の列
- 数を順番に並べたものがベクトルの座標で、次元を増やせば無限に高次元に拡張可能です。

### 3. ベクトルの基本演算
- **スカラー倍 (Scalar Multiplication)**: $k \in \mathbb{R}$ を用いてベクトル $\mathbf{x}$ を $k$ 倍する操作。
- **ベクトル和 (Vector Addition)**: ベクトル同士の加算であり、幾何学的には二つの矢印を平行移動してその和を求めます。

### 4. 個別の成績表と平均の計算
- 個別の成績 $k$ と英語の成績 $e$ が与えられ、これらの平均を計算しています。
  - 例:
$$ k =
\begin{bmatrix}
100 \\
70 \\
30 \\
45 \\
80
\end{bmatrix}
, \quad
e =
\begin{bmatrix}
83 \\
55 \\
35 \\
60 \\
60
\end{bmatrix}
$$

  - 平均 (avg) の計算:
    $$
    \text{avg} = \frac{1}{2} k + \frac{1}{2} e
    $$

### 5. ベクトルの線形結合 (Linear Combination)
- ベクトルの線形結合の例として、$\mathbf{i}$ と $\mathbf{j}$ との組み合わせが示されています。これはベクトル空間において、新たなベクトルを生成するための基本的な手法です。
