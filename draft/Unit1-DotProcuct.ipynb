##内積

**学習目標**：ベクトル積は2種類あり、内積（Inner Product, Scalar Product）と外積（Outer Product, Vector Product）の結果と意味、活用に大きな違いがあることをしっかり理解しましょう。

特に、内積は機械学習において、使用頻度が高く、ベクトル同士の類似度を計算するために使用するコサイン類似度とも密接に関係します。そのあたりも意識して理解を深めるようにしましょう。

任意のベクトル$\mathbf{a}$と$\mathbf{b}$の内積は次式のように定義される。

$$\mathbf{a}\cdot\mathbf{b}=\mathbf{a}^{T}\mathbf{b}=\mathbf{b}^{T}\mathbf{a}$$

$$\mathbf{a}\cdot\mathbf{b}= \|\mathbf{a}\|\|\mathbf{b}\|\cos{\theta} $$

ベクトル$\mathbf{a}$と$\mathbf{b}$が次のように与えられたら

$$ \mathbf{a}= \begin{bmatrix}
 1 \\ 3 \\
\end{bmatrix} , \mathbf{b}= \begin{bmatrix}
 5 \\ 1 \\
\end{bmatrix} $$

結果は次のようになります。

```python
import matplotlib.pyplot as plt
import numpy as np

# ベクトルの定義
a = np.array([1, 3])
b = np.array([5, 1])

# グラフの設定
plt.figure()
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector b')

# 軸の設定
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# ラベルと凡例の設定
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Vectors a and b')

# グラフを表示
plt.show()
```

ベクトル$\mathbf{a}$と$\mathbf{b}$が次のように与えられたら

$$ \mathbf{a}+\mathbf{b}= \begin{bmatrix}
 1 \\ 3 \\
\end{bmatrix} + \begin{bmatrix}
 5 \\ 1 \\
\end{bmatrix} = \begin{bmatrix}
 1+5 \\ 3+1 \\
\end{bmatrix} $$

```python
import matplotlib.pyplot as plt
import numpy as np

# ベクトルの定義
a = np.array([1, 3])
b = np.array([5, 1])

# ベクトルの和を計算
vector_sum = a + b

# グラフの設定
plt.figure()
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector b')
plt.quiver(0, 0, vector_sum[0], vector_sum[1], angles='xy', scale_units='xy', scale=1, color='g', label='Vector a+b')
plt.plot([a[0], vector_sum[0]], [a[1], vector_sum[1]], 'k--')
plt.plot([b[0], vector_sum[0]], [b[1], vector_sum[1]], 'k--')

# 軸の設定
plt.xlim(0, 7)
plt.ylim(0, 7)
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# ラベルと凡例の設定
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Vectors a, b and their Sum (a+b)')

# グラフを表示
plt.show()
```

ベクトル$\mathbf{a}$と$\mathbf{b}$の内積は次式となる。

$$\mathbf{a}\cdot\mathbf{b}=
 \begin{bmatrix} 1 \ 3 \\\end{bmatrix}\begin{bmatrix}  5 \\ 1 \\\end{bmatrix}=1\cdot5+3\cdot1=8 $$

$$\mathbf{a}\cdot\mathbf{b}= \|\mathbf{a}\|\|\mathbf{b}\|\cos{\theta} $$
これを以下のように書き換えると

$$\mathbf{a}\cdot\mathbf{b}= \|\mathbf{a}\|\cos{\theta}\|\mathbf{b}\| $$
これはベクトル$\mathbf{a}$をベクトル$\mathbf{b}$に射影(projection)したものと$\mathbf{b}$のノルムの積を意味する。逆に

$$\mathbf{a}\cdot\mathbf{b}= \|\mathbf{b}\|\cos{\theta}\|\mathbf{a}\| $$

ベクトル$\mathbf{b}$をベクトル$\mathbf{a}$に射影(projection)したものと$\mathbf{a}$のノルムの積を意味する。

```python
import matplotlib.pyplot as plt
import numpy as np

# ベクトルの定義
a = np.array([1, 3])
b = np.array([5, 1])

# ベクトル a を b に射影する
projection_ab = (np.dot(a, b) / np.dot(b, b)) * b


# グラフの設定
plt.figure()

# ベクトル a, b, 射影ベクトルを描画
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector b')
plt.quiver(0, 0, projection_ab[0], projection_ab[1], angles='xy', scale_units='xy', scale=1, color='g', label='Projection of a onto b')

# 射影ベクトルからベクトルaの先端までの垂線を描画
plt.plot([projection_ab[0], a[0]], [projection_ab[1], a[1]], 'k--', label='Perpendicular from projection to a')

# 軸の設定
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# ラベルと凡例の設定
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Projection of Vector a onto Vector b with Perpendicular Line')

# グラフを表示
plt.show()
```

上図からわかるように、図において赤で表されているベクトル$\mathbf{b}$を青で表されているベクトル$\mathbf{a}$に射影して得られる長さ$\|\mathbf{b}\|\cos{\theta}$とベクトル$\mathbf{a}$の長さ$\|\mathbf{b}\|$の積になります。


ベクトル$\mathbf{a}$とベクトル$\mathbf{a}$の内積は
$$\mathbf{a}\cdot\mathbf{a}= \|\mathbf{a}\|\|\mathbf{a}\|= \|\mathbf{a}\|^2 $$

となり、また、ベクトル$\mathbf{a}$の大きさは、次式となる。
$$  \|\mathbf{a}\|= \sqrt{\|\mathbf{a}\|^2} = \sqrt{\mathbf{a}\cdot\mathbf{a}} = \sqrt{\mathbf{a}^{T}\mathbf{a}} $$

###単位ベクトル（Unit vector）

大きさ1のベクトル

任意のベクトルを自分自身の大きさで割ると、方向は同じで大きさが1のベクトルが得られる。⇒Normalization

$$ \frac{\mathbf{a}}{\sqrt{\mathbf{a}^{T}\mathbf{a}}} $$

自分を自分の大きさで割ると単位長さの同じ方向のベクトルが残ります。
