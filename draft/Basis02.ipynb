## 基底(Basis)

### 新しい座標系と基底の導入
1. **基底ベクトルの定義:**
   - 標準座標系では、基底ベクトルは $\mathbf{e}_x = (1, 0)$ と $\mathbf{e}_y = (0, 1)$ です。これらがデフォルトの基底ベクトル（標準基底）です。
   - 点 $(2, 3)$ は標準基底で表すと、 $2\mathbf{e}_x + 3\mathbf{e}_y$ となります。

2. **新しい基底の導入:**
   - 新しい基底 $ B $ を $ \begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix} $ と定義しています。これにより、新しい座標系が定義されます。
   - 新しい基底を使うと、同じベクトルを異なる座標として表すことができます。

3. **新しい基底を用いた座標表現:**
   - 例として、元の座標 $(2, 2)$ を新しい基底 $ B $ の座標系で表現する過程が示されています。
   - 新しい基底での表現は次のように行われます:
   $$ \begin{bmatrix} 2 \\ 2 \end{bmatrix} = k_1 \begin{bmatrix} 1 \\ 1 \end{bmatrix} + k_2 \begin{bmatrix} 1 \\ -1 \end{bmatrix} $$
   - これを行列形式で書くと、
   $$ \begin{bmatrix} 2 \\ 2 \end{bmatrix} = \begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix} \begin{bmatrix} k_1 \\ k_2 \end{bmatrix} $$
   
   - これを解くことで、新しい基底での座標 $\begin{bmatrix} k_1 \\ k_2 \end{bmatrix}$ が得られます。

### まとめ
- 新しい基底を導入することで、同じベクトルや点を異なる視点から表現できます。これにより、座標系を変換して問題を解決する際に便利です。
- このページの内容は、座標変換や異なる基底でのベクトルの表現方法についての理解を深める助けとなります。


このページでは、標準座標系でのベクトルの表現を新しい基底 $ C $ に基づく座標系に変換する方法と、任意の基底間の変換について説明しています。

### 1. 基準座標系から新しい基底への変換
- **標準座標系のベクトル $ X $**:
  $$
  X = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}_{\mathbf{e}}
  $$
  ここで、$\mathbf{e}$ は標準基底を表します。

- **新しい基底 $ C $**:
  $$
  C = \begin{bmatrix} \mathbf{c}_1 & \mathbf{c}_2 \end{bmatrix}
  $$
  ここで、$\mathbf{c}_1$ と $\mathbf{c}_2$ は新しい基底ベクトルです。

- **$ C $ における座標**:
  ベクトル $ X $ を基底 $ C $ で表現すると、その座標 $ Y $ は次のようになります。
  $$
  Y = \begin{bmatrix} y_1 \\ y_2 \end{bmatrix}_C
  $$
  これにより、標準基底での表現が新しい基底 $ C $ に対応する座標系に変換されます。

### 2. 任意の基底間の変換
- **基底 $ B $, $ C $** の設定:
  - 基底 $ B = \{\mathbf{v}_1, \mathbf{v}_2\} $
  - 基底 $ C = \{\mathbf{w}_1, \mathbf{w}_2\} $

- **基底間の変換式**:
  ベクトル $ \mathbf{v}_2 $ を基底 $ C $ の組み合わせで表現します。ベクトル $ V $ を次のように変換します。
  $$
  V = l_1 \mathbf{w}_1 + l_2 \mathbf{w}_2
  $$
  この表現は、基底 $ B $ の組み合わせから基底 $ C $ の組み合わせに変換するプロセスを示しています。

- **具体的な変換手順**:
  $$
  \begin{bmatrix} \mathbf{v}_1 & \mathbf{v}_2 \end{bmatrix} = \begin{bmatrix} a & b \\ c & d \end{bmatrix} \begin{bmatrix} \mathbf{w}_1 & \mathbf{w}_2 \end{bmatrix}
  $$
  これにより、ベクトルの座標が基底 $ C $ でどのように表されるかが明示されます。

### まとめ
- 新しい基底への変換は、座標系を変更し、ベクトルの視点を変えるために重要です。
- 基底間の変換は、異なる基準から同じベクトルを見た場合の関係性を示し、線形代数における多くの応用につながります。


このページでは、基底変換行列、連立方程式の解法、行列の操作について説明しています。

### 1. 基底変換行列 (Transition Matrix)
- 行列 $\begin{bmatrix} a & c \\ b & d \end{bmatrix}$ は基底変換行列を表しています。この行列は、ある基底から別の基底への変換を行います。

### 2. 連立方程式と行列
- 連立方程式の例として、次の2つの式が与えられています:
  1. $2x + 3y = 1$
  2. $4x + 6y = 3$

- この方程式を解くための手順が示されています。1つ目の方程式から $ y = 1 $ を代入して $ x = -1 $ を求めています。

### 3. 行列を用いた解法
- これらの連立方程式は、行列を使って次のように表すことができます:
  $$
  A = \begin{bmatrix} 2 & 3 \\ 4 & 6 \end{bmatrix}, \quad \mathbf{x} = \begin{bmatrix} x \\ y \end{bmatrix}, \quad \mathbf{b} = \begin{bmatrix} 1 \\ 3 \end{bmatrix}
  $$
  連立方程式は行列方程式 $ A\mathbf{x} = \mathbf{b} $ となります。

- 行列 $ [A|\mathbf{b}] = \begin{bmatrix} 2 & 3 & | & 1 \\ 4 & 6 & | & 3 \end{bmatrix} $ を用いて、ガウス消去法などで解を求めます。

### 4. 行列の基本操作
- 行列の基本操作には以下の3つがあります:
  1. **Row multiplication (行の掛け算)**: 行の要素に定数をかける。
  2. **Row switching (行の入れ替え)**: 2つの行を入れ替える。
  3. **Row addition (行の加算)**: ある行に他の行の定数倍を足す。

- 例として、単位行列 $ I = \begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix} $ から、行の掛け算で基本行列 $ E $ を生成する方法が示されています。基本行列 (elementary matrix) は行の操作を表現するために使用されます。

### まとめ
- 基底変換行列を用いることで、異なる基底間での座標変換が可能になります。
- 行列を使った連立方程式の解法は、行の操作を行うことで簡便に解を求めることができます。
- 行列の基本操作は、行列計算の基礎であり、様々な応用で利用されます。
