
このページでは、行列 $ A = \begin{bmatrix} 2 & 1 \\ 4 & 2 \end{bmatrix} $に関する「行空間（row space）」、「列空間（column space）」、「ゼロ空間（null space）」について説明します。

### 1. 行空間（Row Space）
- 行空間は、行列 \( A \) の行ベクトルによって張られる空間です。
- 行列 \( A \) の行ベクトル \([2, 1]\) と \([4, 2]\) は同じ直線上にあります。この行ベクトルたちが張る空間が行空間です。
- 図で青色の直線として示されています。

### 2. 列空間（Column Space）
- 列空間は、行列 \( A \) の列ベクトルによって張られる空間です。
- \( A \) の列ベクトルは \([2, 4]\) と \([1, 2]\) です。これらのベクトルによって張られる平面が列空間です。
- 図では赤色の直線として表されています。列空間はこの赤い直線が \( (2,4) \) と \( (1,2) \) のベクトル方向を示しています。

### 3. ゼロ空間（Null Space）
- ゼロ空間とは、行列 \( A \) の解がゼロベクトルになるようなベクトルの集合です。
- 数式で表すと、ゼロ空間の要素 \(\mathbf{x}\) は、 \( A\mathbf{x} = \mathbf{0} \) を満たすものです。
- 図では、ゼロ空間は青い線で示されており、これは元の空間からゼロへのマッピングを表しています。

### まとめ
- 行空間と列空間は行列 \( A \) によって張られる空間であり、行列の特性を視覚的に理解するために重要です。
- ゼロ空間は、行列が作用する際にゼロにマッピングされるベクトルの集合であり、行列の可逆性や解の存在を考える上で重要な概念です。

このノートは、線形代数における行列の持つ空間の構造を視覚的に説明しており、行列の作用を理解する助けとなります。
