### 2) 行列のかけ算の性質

#### i) スカラー倍

行列のスカラー倍は次のように計算されます：

$$
(A B) = A (c B) = c (A B)
$$

ここで、  
$ A = m \times n $,  
$ B = n \times k $,  
$ c \in \mathbb{R} $,  
$ C = n \times 1 $.

$$
C (A B)_{ij} = C r_i \cdot C_j
$$

行列 $ A $ と $ B $ が次のように定義される場合：

$$
A =
\begin{bmatrix}
r_1 \\
r_2 \\
\vdots \\
r_m
\end{bmatrix}, \quad
B =
\begin{bmatrix}
c_1 & c_2 & \cdots & c_k
\end{bmatrix}
$$

スカラー倍の行列 $ c A $ は：

$$
c A =
\begin{bmatrix}
c r_1 \\
c r_2 \\
\vdots \\
c r_m
\end{bmatrix}
$$

$(cA)B_{ij} = c r_i \cdot c_j$

行列 $ C B $ も次のように表せます：

$$
C B =
\begin{bmatrix}
c_1 c_2 \cdots c_k
\end{bmatrix}, \quad
(A (C B))_{ij} = r_i \cdot c_j
$$

行列の積の項は：

$$
AB_{ij} = (a_{i1}, a_{i2}, \ldots, a_{in}) \cdot (b_{1j}, b_{2j}, \ldots, b_{nj}) = \sum_{s=1}^{n} a_{is} b_{sj}
$$

拡張すると：

$$
(AB)_{ij} = \sum_{s=1}^{n} a_{is} b_{sj}, \quad (AB B_k)_{ij} = \sum_{s=1}^{n} a_{is} b_{sj} c_{sk}
$$

---

### ii) 行列のかけ算の性質 (2)

行列の積についての性質です：

$$
A (BC)_{ij} = \sum_{s=1}^{n} a_{is} b_{sj}
$$

ここで、  
$ B = n \times k $,  
$ (BC)_{ij} = \sum_{s=1}^{n} b_{is} c_{sj} $

次に、行ベクトルとして次のように示されます：

$$
(BC)_{ij} = \sum_{s=1}^{n} b_{is} c_{sj}
$$

行列 $ A $ と行ベクトル $ (a_{i1}, a_{i2}, \ldots, a_{in}) $ をかけた場合：

$$
(ABC)_{ij} = \sum_{s=1}^{n} a_{is} b_{sj} c_{sj} = \sum_{s=1}^{n} a_{is} b_{sj} c_{sk}
$$

これにより行列の積の順序に応じた性質が示されます。


### 3) 行列の分配法則

行列の分配法則は次のようになります：

$$
A(B + C) = AB + AC
$$

$$
(A + B)C = AC + BC
$$

行列の積は一般には可換ではありません：

$$
AB \neq BA
$$

#### 証明

行列 $ A $, $ B $, $ C $ が以下のサイズの場合：

- $ A: m \times n $
- $ B: n \times k $
- $ C: n \times k $

行列の具体例：

$$
A =
\begin{bmatrix}
a_1 \\
a_2 \\
\vdots \\
a_m
\end{bmatrix}, \quad
B =
\begin{bmatrix}
b_1, b_2, \ldots, b_k
\end{bmatrix}, \quad
C =
\begin{bmatrix}
c_1, c_2, \ldots, c_k
\end{bmatrix}
$$

$(A (B + C))_{ij}$ の計算：

$$
(A (B + C))_{ij} = a_i \cdot (b_j + c_j)
$$

行列の要素は次のようになります：

$$
(B + C)_{j} = (b_{1j} + c_{1j}, b_{2j} + c_{2j}, \ldots, b_{kj} + c_{kj})
$$

このため、

$$
(A (B + C))_{ij} = a_i \cdot (b_j + c_j) = a_i b_j + a_i c_j = (AB)_{ij} + (AC)_{ij}
$$

### 13) 単位行列 (Identity Matrix)

#### 1) 単位行列の定義

単位行列 $ I $ は、対角成分がすべて 1 であり、それ以外の成分がすべて 0 の行列です。

例：

$$
I_2 =
\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}, \quad
I_3 =
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
$$

#### 2) 行列の積における列の分解

$$
AB = C
$$

$$
AB_1 = c_1, \quad AB_2 = c_2, \quad \ldots, \quad AB_n = c_n
$$

#### 3) 単位行列の性質

##### i) $ IA = A $

$$
IA \to I: n \times n, \quad A: n \times k
$$

$$
\mathbf{y} = I \mathbf{x} \to
\begin{bmatrix}
y_1 \\
y_2 \\
\vdots \\
y_n
\end{bmatrix}
=
\begin{bmatrix}
1 & 0 & \ldots & 0 \\
0 & 1 & \ldots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \ldots & 1
\end{bmatrix}
\begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{bmatrix}
$$

単位行列をかけることで、元の行列やベクトルがそのままの形で得られることが確認できます。

### 3) 単位行列の性質

#### ii) $ AI = A $ か？

$$
I_{n \times n} =
\begin{bmatrix}
e_1 & e_2 & \ldots & e_n
\end{bmatrix}
$$

$$
A I = A, \quad I A = I A = A
$$

### 24) 逆行列 (Inverse Matrix)

#### 1) 定義

行列 $ A $ に対して、$ A X = X A = I $ となる行列 $ X $ が存在すれば、$ A $ は正則行列 (invertible matrix) といい、$ X = A^{-1} $ は $ A $ の逆行列です。

---

#### 2) 逆行列の簡単な性質

行列 $ A $ の逆行列を $ B $, 行列 $ C $ とすると：

$$
AB = BA = I
$$

$$
AC = CA = I
$$

---

#### 3) ゼロ除数要因 (Zero Divisor)

$$
AB = 0 \implies \text{正則行列ではない}
$$

具体例：

$$
\begin{bmatrix}
1 & 1 \\
1 & 1
\end{bmatrix}, \quad
\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}, \quad
\begin{bmatrix}
0 & 0 \\
0 & 0
\end{bmatrix}
$$

---

#### 4) $ 2 \times 2 $ 行列の逆行列

行列 $ A =
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
$ の逆行列は：

$$
A^{-1} = \frac{1}{ad - bc}
\begin{bmatrix}
d & -b \\
-c & a
\end{bmatrix}
$$

ただし、$ ad - bc \neq 0 $ の場合に限り逆行列が存在する。

---

### 5) 可逆行列 (Invertible Matrix)

行列式 (determinant) について：

$$
\text{行列 } A \text{ が可逆} \iff \det(A) = ad - bc \neq 0
$$

---

### 6) 非可逆行列 (Singular Matrix)

行列 $ A $ の行列式が 0 の場合：

$$
\det(A) = ad - bc = 0
$$

この場合、逆行列は存在しません。
