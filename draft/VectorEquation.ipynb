## ベクトル方程式

### 直線の方程式

- **方程式の基本形**:
$
  y = ax + b
$

  - $ a $ は傾き、$ b $ は切片。
  - $b = 0 $ の場合、直線は原点を通る。

- **傾きと直線の関係**:
  
  - 点 $ (1, a) $は、傾きが $ a $ である直線のポイントとなる。
  - この点をベクトルとして、$\mathbf{v} = (1, a) $ と定義。

- **ベクトル方程式の表現**:
  
  - ベクトル $ \mathbf{v} $ をスカラー $t $ で乗算して点の軌跡を表現：
  
    $
    \mathbf{x} = t\mathbf{v}
    $

### ベクトルの方程式と平行移動

- 直線の別の表現方法として、平行移動を考慮に入れる：
  
$
  X = \mathbf{V}_0 + t\mathbf{V}_1
$

  - ここで、$ \mathbf{V}_0 $は初期点（平行移動の始点）、$ \mathbf{V}_1 $は方向ベクトルを示す。
  
- **幾何的解釈**:
  
  - 平行移動を行うと、直線の任意の点をベクトル形式で表すことができる。
  - 図示して、$X = \mathbf{V}_0 + t\mathbf{V}_1 $ という形で表される。

### 平面のベクトル方程式

#### 1. 直線上のベクトル表現

直線上のすべての点を$(\mathbf{v}_1)$で表現できるが、もう1つのベクトルが必要。

$
(1, 2) = 3\mathbf{e}_1 + 2\mathbf{e}_2 \quad (\mathbf{e}_1 = (1, 0), \ \mathbf{e}_2 = (0, 1))
$

$
(a, b) = a\mathbf{e}_1 + b\mathbf{e}_2
$

- $(\frac{\mathbf{v}_1}{\mathbf{v}_2} \implies \mathbf{v}_1 = k\mathbf{v}_2)$

- $((a, b) = a\mathbf{v}_1 + b\mathbf{v}_2 = (a + b)\mathbf{v}_2)$ （直線）

#### 2. 平面のベクトル方程式

平面全体を表現するためには、互いに平行でない2つのベクトルが必要。

$
\mathbf{X} = \mathbf{v}_1 + t\mathbf{v}_2 + t\mathbf{v}_3
$

（2次元 → 2変数）

#### 例1：直線

$y = 3x + 2 $ のベクトル方程式？

- ベクトル $\mathbf{v}_1$ の始点 $(0, 0)$ と終点 $(1, 5)$
- ベクトル $\mathbf{v}_2$ の始点 $(1, 5)$ と終点 $(2, 8)$

$
\mathbf{X} = \mathbf{v}_1 + t\mathbf{v}_2 = (1, 5) + t(2 - 1, 8 - 5) = (1, 5) + t(1, 3)
$

### ベクトル方程式と再パラメータ化

#### 平面のベクトル方程式

1. ベクトルの組み合わせで平面上の任意の点を表現

$
\mathbf{X} = \mathbf{v}_1 + t_0\mathbf{v}_2 + t_2\mathbf{v}_3
$

- $\mathbf{v}_1 = (1, 1, \frac{1}{2})$
- $\mathbf{v}_2 = (0, 0, \frac{5}{2}) - (1, 1, \frac{1}{2}) = (-1, -1, 1)$
- $\mathbf{v}_3 = (0, 0, \frac{3}{2}) - (1, 1, \frac{1}{2}) = (-1, 0, \frac{1}{3})$

よって、

$
\mathbf{X} = (1, 1, \frac{1}{2}) + t_0(-1, -1, 1) + t_1(-1, 0, \frac{1}{3})
$

展開して、

$
\mathbf{X} = (1 - t_0 - t_1, 1 - t_0, \frac{1}{2} + t_0 + \frac{1}{3}t_1)
$

#### 2. 再パラメータ化 (Reparameterization)

- ベクトル方程式の初点

1. $ \mathbf{v}_1 = (1, 1) $
2. $ \mathbf{v}_2 = (2, 2) $

- $ x = t\mathbf{v}_1 $ または $ (t, t) $
- $ g(t) = t\mathbf{v}_2 = t(2, 2) \rightarrow (2t, 2t) $

同じ直線を表すが、速さが異なる。

#### 3. 再パラメータ化されたベクトル方程式との違い

- パラメータ $t$ (time) の増加に伴い速さが変わる。

直線速度 $v$ は変数 $t = 0 \rightarrow$ 速度が変わった

例えば、直線 $y = x$ のベクトル方程式

$
f(t) = (t, t)
$

の場合、時刻 $t$ に対する方向の変換が可能。

- $f'(t) = (1, 1)$: 瞬間的変化率 → 速度 = 距離 / 時間

$
h(t) = \frac{f(t)}{\sqrt{2}}
$

- 速度が1に統一される。
