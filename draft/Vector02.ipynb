### ベクトルとは
1. **ベクトルの定義**:
   - ベクトルは「矢印のようなもの」で、向きと大きさを持つ量を表します。
   - 物理的な特性（力や速度など）を反映するため、座標系の変換に対して不変であることが理想とされていますが、実際のベクトルの成分は座標系の変換により変化することもあります。

2. **数値の並びとしてのベクトル**:
   - ベクトルは数値（成分）を順に並べたものとして考えられます。
   - 成分数が次元数を決め、ベクトルの表記法が座標変換によって変化します（「不変」ではない）。

3. **ベクトルと空間**:
   - ベクトルはベクトル空間の要素であり、数学的にきちんと定義された構造を持っています。
   - ベクトル同士の足し算やスカラー倍（数を掛ける操作）などの演算が定義されています。

4. **ベクトルの基本演算**:
   - **スカラー倍**: スカラー $ k $ とベクトル $ x $ に対して、 $ kx $ という新しいベクトルが定義されます。
   - **ベクトルの和**: 2つのベクトルの和 $ \alpha + \beta $ は、図で示されるように矢印を平行移動させて合成したものとして視覚的に理解できます。

5. **データのベクトル表現**:
   - 多くのデータ（数値リスト、画像、音声、テキストなど）はベクトルとして表現できます。
   - これにより、ベクトルの基本演算がデータ操作の基本機能となるため、ベクトルの理解は非常に重要です。


### ベクトルの線形結合とベクトル空間の形成
- ベクトルの線形結合（例: 図中のベクトル）は、ベクトル空間の基本的な構成要素です。
- 複数のベクトルの組み合わせ（線形結合）によって、新たなベクトルを生成し、ベクトル空間を形成します。

### 行列の本質
- 行列は、関数の役割を果たし、ベクトルを他のベクトルへと「写像（mapping）」します。
- 例えば、行列 $ f = \begin{bmatrix} a & b \\ c & d \end{bmatrix} $ は入力ベクトルに対して、出力ベクトルを生成します。

#### 行列の積とその結果
- 行列 $ f $ と $ g $ の積は、各要素を計算して新しい行列を生成します。
- 行列の積はベクトルの変換を表し、行列同士の掛け算が対応するベクトルの座標を変更する処理になります。

### 行列の積と内積
- 行列の積はベクトルの内積（inner product）に繋がり、2つのベクトル間の相互作用（類似度）を計算できます。
- 行ベクトルと列ベクトルの積により、スカラー値が得られます（内積の計算）。

#### 行列積の具体例
- 例えば、行ベクトル $ \begin{bmatrix} 1 & 3 \end{bmatrix} $ と列ベクトル $ \begin{bmatrix} a \\ c \end{bmatrix} $ の積は $ 1 \times a + 3 \times c $ となり、これが内積の計算を意味します。

- 列ベクトルの組み合わせの例として、 $ a \begin{bmatrix} 1 \\ 3 \end{bmatrix} + c \begin{bmatrix} 2 \\ 4 \end{bmatrix} $ が示されており、2つの引数ベクトルの線形結合を表現しています。
