## 固有値と固有ベクトル

### 固有値と固有ベクトルの概要
- 固有値と固有ベクトルは、行列 $ A $ がベクトル $ \mathbf{x} $ を変換したとき、その方向を変えずにスケール（大きさ）のみを変えるベクトルとそのスケールの値を指します。
- 数式で表すと、次のようになります：
  $$
  A\mathbf{x} = \lambda \mathbf{x}
  $$
  ここで、$ \lambda $ は固有値、$ \mathbf{x} $ は固有ベクトルです。

### 固有値の求め方
1. 行列 $ A $ の固有値は次の式から求めます：
   $$
   (A - \lambda I)\mathbf{x} = 0
   $$
   ここで、$ I $ は単位行列です。

2. 上記の方程式が非自明解（$\mathbf{x} \neq 0$）を持つためには、行列 $ B = (A - \lambda I) $ が正則でない、つまり $ \text{det}(B) = 0 $ である必要があります。

3. この条件を用いて固有値 $ \lambda $ を求めます。

#### 例：行列 $ A = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix} $ の固有値と固有ベクトル
1. 固有値を求めるために次の式を立てます：
   $$ \text{det}(A - \lambda I) = 0 $$


   $$ \text{det} \begin{bmatrix} 2 - \lambda & 1 \\ 1 & 2 - \lambda \end{bmatrix} = (2 - \lambda)^2 - 1 = 0 $$


   $$ (\lambda - 1)(\lambda - 3) = 0 $$

   よって、固有値は $ \lambda_1 = 1 $ と $ \lambda_2 = 3 $ です。

2. 各固有値に対応する固有ベクトルを求めます。

   - 固有値 $ \lambda_1 = 1 $ の場合：
   $$ (A - I)\mathbf{x} = \begin{bmatrix} 1 & 1 \\ 1 & 1 \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} $$

これを解くと、$ x_1 = -x_2 $。したがって、固有ベクトルは $ \mathbf{x} = c_1 \begin{bmatrix} -1 \\ 1 \end{bmatrix} $。

   - 固有値 $ \lambda_2 = 3 $ の場合：

$$ (A - 3I)\mathbf{x} = \begin{bmatrix} -1 & 1 \\ 1 & -1 \end{bmatrix}
\begin{bmatrix} x_1 \\ x_2 \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} $$


これを解くと、$ x_1 = x_2 $。したがって、固有ベクトルは $ \mathbf{x} = c_2 \begin{bmatrix} 1 \\ 1 \end{bmatrix} $。

### まとめ
- 固有値と固有ベクトルは、行列がどのようにベクトルを変換するかの特性を示します。特に、固有値はベクトルの拡大・縮小の倍率を、固有ベクトルはその変換で不変の方向を示します。
- 固有値を求めるには、行列の特性方程式を解く必要があり、その結果から固有ベクトルも求められます。



### 関数の定義### 固有ベクトルに関する計算
- 行列 $ A $ に対して、固有値 $ \lambda $ と固有ベクトル $ \mathbf{x} $ の関係式は次のように表されます。
  $$
  A\mathbf{x} = \lambda \mathbf{x}
  $$
- 具体的な例で、固有ベクトルの確認が行われています。
  - 固有値 $ \lambda = 1 $ の場合、対応する固有ベクトルとして $ \mathbf{x} = \begin{bmatrix} 1 \\ -1 \end{bmatrix} $ が求められています。
  - 固有値 $ \lambda = 3 $ の場合、対応する固有ベクトルとして $ \mathbf{x} = \begin{bmatrix} 1 \\ 1 \end{bmatrix} $
