### 行列式の性質

1. $\text{det}(\mathbf{A}) = 0$ のとき、$\mathbf{A}$ は singular。
2. 行列が Rank-deficient であることと、$\text{det}(\mathbf{A}) = 0$ であることは同値。
3. 対角行列の場合：
$ \text{det}(\mathbf{A}) = a_{11}a_{22}a_{33} \ldots a_{nn} $

4. 三角行列の場合も同様に、対角成分の積となる：
$ \text{det}(\mathbf{A}) = a_{11}a_{22}a_{33} \ldots a_nn} $
このように行列式の計算や逆行列の性質

### ※三角行列

- **上三角行列 (Upper Triangular Matrix)** と **下三角行列 (Lower Triangular Matrix)** の例：
$$ \begin{bmatrix}
4 & 3 \\
0 & -2
\end{bmatrix},
\begin{bmatrix}
2 & 0 & 0 \\
-3 & 5 & 0 \\
6 & 4 & 6
\end{bmatrix},
\begin{bmatrix}
1 & 3 & 4 & 2 \\
0 & 5 & 6 & 9 \\
0 & 0 & 3 & 6 \\
0 & 0 & 0 & 7
\end{bmatrix} $$

### 行列式 (Determinant) の性質

1. 単位行列の行列式は $\text{det}(\mathbf{I}) = 1$
2. $\text{det}(c\mathbf{A}) = c^n \text{det}(\mathbf{A})$（$n \times n$ 行列の場合）
3. $\text{det}(\mathbf{A}^T) = \text{det}(\mathbf{A})$
4. $\text{det}(\mathbf{AB}) = \text{det}(\mathbf{A}) \text{det}(\mathbf{B})$
5. $\text{det}(\mathbf{A}^{-1}) = \frac{1}{\text{det}(\mathbf{A})}$
6. $\text{det}(\mathbf{A}) = \lambda_1 \lambda_2 \ldots \lambda_n$（固有値の積）

### トレース (Trace) の性質

- **トレース (trace)** とは、行列の対角成分の和です。
$ \text{tr}(\mathbf{A}) = \sum_{i=1}^{n} a_{ii} $

トレースの主な性質：

1. $\text{tr}(\mathbf{A} + \mathbf{B}) = \text{tr}(\mathbf{A}) + \text{tr}(\mathbf{B})$
2. $\text{tr}(c\mathbf{A}) = c\text{tr}(\mathbf{A})$
3. $\text{tr}(\mathbf{A}^T) = \text{tr}(\mathbf{A})$
4. $\text{tr}(\mathbf{AB}) = \text{tr}(\mathbf{BA})$ （$\mathbf{A}: m \times n$, $\mathbf{B}: n \times m$）
5. $\text{tr}(\mathbf{A}\mathbf{B}^T) = \text{tr}(\mathbf{B}\mathbf{A}^T)$
6. $\text{tr}(\mathbf{ABCD}) = \text{tr}(\mathbf{BCDA}) = \text{tr}(\mathbf{CDAB}) = \text{tr}(\mathbf{DABC})$（循環性の性質）

行列 $\mathbf{A}: m \times n$, $\mathbf{D}: n \times m$ の場合：
$ \text{tr}(\mathbf{A}) = \sum_{i=1}^{n} \lambda_i $
