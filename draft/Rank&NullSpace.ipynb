### 行列の階数 (Rank)

- **行列の階数 (rank)**：行列が持つ線形独立な列の数または行の数。  
  **Column Spaceの次元** (列空間の次元) と **Row Spaceの次元** (行空間の次元) を意味します。

$$
\text{independentな columnの数} = \text{independentな rowの数}
$$

$$
\text{rank}(\mathbf{A}) = \text{rank}(\mathbf{A}^T)
$$

行列の例：

$$
\mathbf{A} =
\begin{bmatrix}
1 & 2 & 3 \\
0 & 0 & 0
\end{bmatrix}_{2 \times 3}, \quad \text{rank} = 1 \quad (\text{rank-deficient, 希薄})
$$
<br>
$$
\mathbf{A} =
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0
\end{bmatrix}_{2 \times 3}, \quad \text{rank} = 2 \quad (\text{full row rank})
$$

行列のサイズとrank：

- $3 \times 2$ の行列では $\text{rank} = 2$ (full column rank)
- $3 \times 3$ の行列では $\text{rank} = 3$ (full rank)
- $\text{rank} < 2$ の場合は rank-deficient

### Null Space

**Null Space** は方程式 $\mathbf{A}\mathbf{x} = \mathbf{0}$ を満たすベクトル $\mathbf{x}$ の集合。

例：

$$
\mathbf{A} =
\begin{bmatrix}
1 & 0 & 1 \\
0 & 1 & 1
\end{bmatrix}_{2 \times 3}
$$

$$
\mathbf{A}\mathbf{x} = x_1
\begin{bmatrix}
1 \\
0
\end{bmatrix}
+ x_2
\begin{bmatrix}
0 \\
1
\end{bmatrix}
+ x_3
\begin{bmatrix}
1 \\
1
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0
\end{bmatrix}
$$

$$
\mathbf{x} =
\begin{bmatrix}
0 \\
0 \\
0
\end{bmatrix}
$$
←自明解、他に

$$
\mathbf{x} =
\begin{bmatrix}
1 \\
1 \\
-1
\end{bmatrix}
、
\mathbf{x} =
\begin{bmatrix}
2 \\
2 \\
-2
\end{bmatrix}
$$
など解は無数に存在します。よって、$c\mathbf{A}\mathbf{x} = \mathbf{0}$
$$
\mathbf{x}_i = c
\begin{bmatrix}
1 \\
1 \\
-1
\end{bmatrix}
$$
<br>
Null Spaceの次元：

$$
\text{Null Spaceの次元} \to 1, \quad \text{rank} \to 2
$$

例：

$$
\mathbf{A} =
\begin{bmatrix}
1 & 2 & 3 \\
0 & 0 & 0
\end{bmatrix}, \quad \mathbf{x}_1 =
\begin{bmatrix}
-2 \\
1 \\
0
\end{bmatrix}, \quad \mathbf{x}_2 =
\begin{bmatrix}
-3 \\
0 \\
1
\end{bmatrix}
$$

$$
\mathbf{A}(c_1 \mathbf{x}_1) = 0, \quad \mathbf{A}(c_2 \mathbf{x}_2) = 0, \quad \mathbf{A}(c_1 \mathbf{x}_1 + c_2 \mathbf{x}_2) = 0
$$

### Null Spaceと行列の関係

行列 $\mathbf{A}$ が $m \times n$ のとき、Null Spaceの次元は次のように表されます：

$$
\text{dim}(\mathcal{N}(\mathbf{A})) = n - r = 3 - 2 = 1
$$

行列の階数との関係：

$$
\text{dim}(\mathcal{N}(\mathbf{A})) + \text{dim}(\mathcal{R}(\mathbf{A})) = n
$$

この関係は、行列の行空間と列空間が次元 $n$ であることを示しています。

### Null SpaceとLeft Null Space

- Null Space ($\mathcal{N}(\mathbf{A})$) と Left Null Space ($\mathcal{N}(\mathbf{A}^T)$) の関係を示す図。
- $\mathbf{Ax} = 0$ の解空間は $\mathbf{R}^n$ の中にあり、Left Null Space は $\mathbf{R}^m$ に属します。

Null Spaceの次元の公式：

$$
\text{dim}(\mathcal{N}(\mathbf{A})) = m - r, \quad \text{dim}(\mathcal{N}(\mathbf{A}^T)) + \text{dim}(\mathcal{C}(\mathbf{A})) = m
$$

$\mathbf{A}(\mathbf{x}_r + \mathbf{x}_n) = \mathbf{A}\mathbf{x}_r + \mathbf{A}\mathbf{x}_n$ であることから：

$$
\mathbf{A}(\mathbf{x}_r + \mathbf{x}_n) = \mathbf{b}
$$

---

### 方程式 $\mathbf{A}\mathbf{x} = \mathbf{b}$ の解の数

例1：

$$
\begin{cases}
x + 2y = 1 \\
-2x + 4y = 2
\end{cases}
\quad \Rightarrow \text{同じ直線} \rightarrow \text{解は無限}
$$

例2：

$$
\begin{cases}
x + 2y = 1 \\
x + 2y = 2
\end{cases}
\quad \Rightarrow \text{平行直線} \rightarrow \text{解なし}
$$

例3：

$$
\begin{cases}
x + 2y = 1 \\
x + 4y = 1
\end{cases}
\quad \Rightarrow \text{2つの直線が交わる} \rightarrow \text{解が一意に決まる}
$$

---

### Full Column Rankのとき

- **Full Column Rank**: 解が存在しないか、一意に存在する。
- 例として $\mathbf{A}$ が $10 \times 3$ 行列の場合、$\mathbf{b}$ は $\mathbf{C}(\mathbf{A})$ （Column Space）に属する必要があります。

$$
\mathbf{A}
\begin{bmatrix}
x_1 \\
x_2 \\
x_3
\end{bmatrix}
= \mathbf{b}
$$

---

### Full Row Rankのとき

- **Full Row Rank**: 解が一意に決まる。
- Null Spaceが持つ $\mathbf{x}_n$ によって $\mathbf{A}(\mathbf{x}_r + \mathbf{x}_n) = \mathbf{b}$ が表されます。

$$
\mathbf{Ax} = 0 \quad \rightarrow \quad \text{Null Spaceが} \mathbf{x}_n \text{を含む}
$$

---

この関係は、行列のColumn SpaceやNull Spaceに関する重要な情報を含んでいます。特に、行や列のランクが方程式の解の存在や一意性にどのように影響するかが示されています。

### Full Rank (Square Matrix 正方行列) の場合

- **Full Rank** の場合、$\mathbf{Ax} = \mathbf{b}$ の解が一つだけ存在します。

$$
\mathbf{A} =
\begin{bmatrix}
2 & 3 \\
0 & 0
\end{bmatrix}
$$

### Rank-deficient の場合

- $\mathbf{b}$ が $\mathbf{C}(\mathbf{A})$ に属していれば、無限に解が存在します。
- $\mathbf{b}$ が $\mathbf{C}(\mathbf{A})$ に含まれていなければ、解なし。
