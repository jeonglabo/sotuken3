##線形変換
### 1. 線形変換の定義
- 線形変換 $ T $ は以下の性質を持つ場合に線形であると定義されます：
  1. 加法の保存: $ T(\alpha + \beta) = T(\alpha) + T(\beta) $
  2. スカラー倍の保存: $ T(c\alpha) = cT(\alpha) $
  
  これらの性質を満たす変換は「線形変換 (linear transformation)」と呼ばれます。

### 2. 基底ベクトルの変換
- ベクトル $ \begin{bmatrix} x \\ y \end{bmatrix} $ が線形変換 $ T $ によって変換されると、基底ベクトル $ \mathbf{i}, \mathbf{j} $ も新しい基底ベクトル $ \mathbf{i}_{\text{new}}, \mathbf{j}_{\text{new}} $ に変換されます。
- 行列 $ A = \begin{bmatrix} 2 & -3 \\ 1 & 1 \end{bmatrix} $ を用いた線形変換の例が示されています。ベクトル $ \begin{bmatrix} 1 \\ 1 \end{bmatrix} $ が変換され、新しい座標系が生成されています。

### 3. 線形変換の種類
- いくつかの典型的な線形変換の例が示されています：
  - **Shearing** (せん断): $ \begin{bmatrix} 1 & 1 \\ 0 & 1 \end{bmatrix} $
  - **Rotation** (回転): $ 90^\circ $ 回転の場合の行列 $ \begin{bmatrix} \cos \frac{\pi}{2} & -\sin \frac{\pi}{2} \\ \sin \frac{\pi}{2} & \cos \frac{\pi}{2} \end{bmatrix} $
  - **Permutation** (置換): $ x $-軸と $ y $-軸の入れ替え $ \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix} $
  - **Projection on x-axis**: $ \begin{bmatrix} 1 & 0 \\ 0 & 0 \end{bmatrix} $
  - **Projection on a vector**: 特定のベクトルへの射影 $ \begin{bmatrix} 2 & 2 \\ 2 & 4 \end{bmatrix} $
