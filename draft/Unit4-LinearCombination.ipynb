### 線形結合 (Linear Combination)

**学習目標**：線形結合でSpanできる空間の次元、Rank（階数）についてしっかり理解できるようにしましょう。また、行列式（Determinant）との関係、線形独立と従属の関係についても関連付けて考えるようにしましょう。さらに、イメージしにくい3次元以上の空間についてもどのように捉えるへきであるか考えを定めていきましょう。

線形結合の一般形：

$$
a_1 \mathbf{x}_1 + a_2 \mathbf{x}_2 + a_3 \mathbf{x}_3
$$

ベクトルの線形結合による空間のスパン（span）の例：

- 左の図：$\text{span} = 2$
- 右の図：$\text{span} = 1$

Column Spaceの概念：

$$
C(\mathbf{A}) = \text{range}(\mathbf{A}) \quad \text{(Column Space)}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(True, linestyle='--', alpha=0.5)

# Define two linearly dependent vectors (e.g., one is a scalar multiple of the other)
v1 = np.array([2, 1])

# Generate random scalar values to create dependent vectors
np.random.seed(42)  # For reproducibility
scalars = np.random.uniform(-5, 5, 200)

# Compute the points corresponding to the dependent vectors
points = np.array([s * v1 for s in scalars])

# Plot all points that are linearly dependent (along the same line)
ax.scatter(points[:, 0], points[:, 1], color='g', s=10)

plt.show()
```

2行2列の行列の場合、行ベクトル同士または列ベクトル同士が線形独立**(Linear Independence)**であれば、それらの行ベクトル、または列ベクトルによる線形結合は平面にSpanします。すなわち、平面上のすべての点を表現できます。逆に、線形従属であれば、それらによる線形結合は直線にしかSpanできません。

### 線形独立 (Linear Independence)

- 線形独立>直交

$$
a_1 \mathbf{v}_1 + a_2 \mathbf{v}_2 + a_3 \mathbf{v}_3 + \cdots = 0
$$

- ベクトルが直交している場合、それは「orthogonal」であり、独立である (independent) 状態。

### 基底 (Basis)

- 基底とは、ある空間を形成する必須の構成要素。
- 例：2次元の平面を表現する基底。

$$
\begin{bmatrix} 1 \\ 0 \end{bmatrix}, \quad \begin{bmatrix} 0 \\ 1 \end{bmatrix}
$$

$$
\begin{bmatrix} 1 \\ 0 \end{bmatrix}, \quad \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$


でも平面を表現できる⇒ランク2⇒Span=2


しかし、


$$
\begin{bmatrix} 1 \\ 0 \end{bmatrix}, \quad \begin{bmatrix} 2 \\ 0 \end{bmatrix}
$$


では、平面を表現できない⇒ランク1⇒Span=1

特定のベクトルの組み合わせによって表現されないものは、直線や直交座標系と無関係です。

### 単位行列 (Identity Matrix, 単位行列)

単位行列の定義：

$$
\mathbf{I} =
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
$$
<br>
$$
3 \times 1 = 3, \quad \mathbf{A} \times \mathbf{I} = \mathbf{A}
$$

単位行列の具体例：

$$
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{bmatrix}_{2 \times 3}
\times
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}_{3 \times 3}
=
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{bmatrix}_{2 \times 3}
$$

<br>

$$
\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}_{2 \times 2}
\times
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{bmatrix}_{2 \times 3}
=
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{bmatrix}_{2 \times 3}
$$

<br>

$$
\mathbf{A} \mathbf{I} = \mathbf{A}, \quad \mathbf{I} \mathbf{A} = \mathbf{A}
$$
