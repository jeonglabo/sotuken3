### LU分解
LU分解は、行列 $ A $ を下三角行列 $ L $ と上三角行列 $ U $ の積に分解する手法です。

- **下三角行列 $ L $ (Lower Triangular Matrix)**：
  $$
  L =
  \begin{bmatrix}
  l_{11} & 0 & 0 & \cdots & 0 \\
  l_{21} & l_{22} & 0 & \cdots & 0 \\
  l_{31} & l_{32} & l_{33} & \cdots & 0 \\
  \vdots & \vdots & \vdots & \ddots & 0 \\
  l_{n1} & l_{n2} & l_{n3} & \cdots & l_{nn}
  \end{bmatrix}
  $$

- **上三角行列 $ U $ (Upper Triangular Matrix)**：
  $$
  U =
  \begin{bmatrix}
  u_{11} & u_{12} & u_{13} & \cdots & u_{1n} \\
  0 & u_{22} & u_{23} & \cdots & u_{2n} \\
  0 & 0 & u_{33} & \cdots & u_{3n} \\
  \vdots & \vdots & \vdots & \ddots & u_{(n-1)n} \\
  0 & 0 & 0 & \cdots & u_{nn}
  \end{bmatrix}
  $$

### 例題：LU分解による連立方程式の解法
次の連立方程式が与えられています：
$$
\begin{cases}
x + y + z = 6 \\
2x + 3y - z = 5 \\
2x + 3y + 3z = 12
\end{cases}
$$

この連立方程式を解くために、LU分解と行操作を行っています：

1. **行列の初期状態**：
   $$
   \begin{bmatrix}
   1 & 1 & 1 & | & 6 \\
   2 & 3 & -1 & | & 5 \\
   2 & 3 & 3 & | & 12
   \end{bmatrix}
   $$

2. **行操作**：
   - $ r_2 \to r_2 - 2r_1 $ の操作により、次の行列に変換します：
     $$
     \begin{bmatrix}
     1 & 1 & 1 & | & 6 \\
     0 & 1 & -3 & | & -7 \\
     2 & 3 & 3 & | & 12
     \end{bmatrix}
     $$

   - 次に、$ r_3 \to r_3 - r_2 $ でさらに簡単な形に：
     $$
     \begin{bmatrix}
     1 & 1 & 1 & | & 6 \\
     0 & 1 & -3 & | & -7 \\
     0 & 0 & 4 & | & 12
     \end{bmatrix}
     $$

3. **逆代入**で解を求める：
   - $ 4z = 12 $ から $ z = 3 $。
   - 次に、$ y - 9 = -7 $ から $ y = 2 $。
   - 最後に、$ x + 2 + 3 = 6 $ から $ x = 1 $。

この一連の操作により、解 $ x = 1, y = 2, z = 3 $ を求めることができました。

### まとめ
LU分解と行操作を用いることで、連立方程式を系統的に解くことが可能です。この手法は、大規模な行列の計算にも応用可能で、数値計算などで広く使われています。

### LU分解の手順

1. **行操作による行列の変形**:
   - 初期行列：
     $$
     [A|b] =
     \begin{bmatrix}
     1 & 1 & 1 & | & 6 \\
     2 & 3 & -1 & | & 5 \\
     2 & 3 & 3 & | & 12
     \end{bmatrix}
     $$

   - 行操作 $ r_2 \to (r_2 - 2r_1) $ により、次の行列に変形：
     $$
     \begin{bmatrix}
     1 & 1 & 1 & | & 6 \\
     0 & 1 & -3 & | & -7 \\
     2 & 3 & 3 & | & 12
     \end{bmatrix}
     $$

   - 続いて $ r_3 \to (r_3 - r_2) $ の行操作で、上三角行列 $ U $ の形に：
     $$
     \begin{bmatrix}
     1 & 1 & 1 & | & 6 \\
     0 & 1 & -3 & | & -7 \\
     0 & 0 & 4 & | & 12
     \end{bmatrix}
     $$

2. **基本行列の利用によるLU分解**:
   - 基本行列 $ E $ を用いて各行操作を逆行列で表し、行列の分解を進めます。

   - 行交換の基本行列 $ P_{31} $ などを使用し、行列 $ A $ を次のように分解します：
     $$
     A =
     \begin{bmatrix}
     1 & 1 & 1 \\
     2 & 3 & -1 \\
     2 & 3 & 3
     \end{bmatrix}
     =
     \begin{bmatrix}
     1 & 0 & 0 \\
     2 & 1 & 0 \\
     2 & 0 & 1
     \end{bmatrix}
     \begin{bmatrix}
     1 & 1 & 1 \\
     0 & 1 & -3 \\
     0 & 0 & 4
     \end{bmatrix}
     = LU
     $$

   - 下三角行列 $ L $ と上三角行列 $ U $ をそれぞれ識別できます。

3. **行列の分解結果**:
   - $ L $ は下三角行列、$ U $ は上三角行列としてそれぞれ示されています。これにより、元の行列 $ A $ は $ A = LU $ という形で分解され、LU分解が完了します。

### PLU分解の手順

1. **PLU分解の導入**:
   - PLU分解では、行列 $ A $ を置換行列 $ P $、下三角行列 $ L $、上三角行列 $ U $ の積として分解します。
   - 例として、行列 $ A = \begin{bmatrix} 0 & 0 & 3 \\ 2 & 3 & -1 \\ 2 & 3 & 1 \end{bmatrix} $ を考えます。

2. **行操作による分解**:
   - まず、置換行列 $ P_{13} $ を用いて、行を入れ替えます。
     $$
     P_{13}A = \begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix} \begin{bmatrix} 0 & 0 & 3 \\ 2 & 3 & -1 \\ 2 & 3 & 1 \end{bmatrix} = \begin{bmatrix} 2 & 3 & -1 \\ 0 & 0 & 3 \\ 2 & 3 & 1 \end{bmatrix}
     $$
   - 行操作 $ r_2 \to r_2 - \frac{1}{2} r_1 $ によって上三角行列 $ U $ の形に変形します。
   - 結果として、$ A $ は次のように分解されます：
     $$
     A = P_{13}LU = \begin{bmatrix} 0 & 0 & 1 \\ 1 & 0 & 0 \\ 0 & 1 & 0 \end{bmatrix} \begin{bmatrix} 1 & 0 & 0 \\ \frac{1}{2} & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix} \begin{bmatrix} 2 & 3 & -1 \\ 0 & 0 & 3 \\ 0 & \frac{1}{2} & \frac{3}{2} \end{bmatrix}
     $$

### LU分解の利用法

1. **連立方程式の解法**:
   - 与えられた連立方程式 $ Ax = b $ を LU分解を用いて解きます。
   - 方程式を $ LUx = b $ と置き換え、まず $ L $ の逆行列を用いて $ c = L^{-1}b $ を求めます。
   - 次に、上三角行列 $ U $ を用いて $ Ux = c $ の形にし、後退代入により $ x $ を解きます。

2. **例**:
   - 行列 $ A = \begin{bmatrix} 1 & 1 & 1 \\ 2 & 3 & -1 \\ 2 & 3 & 3 \end{bmatrix} $ を用いたLU分解では、次のような形で連立方程式を解く手順が示されています：
     $$
     Ax = b \rightarrow LUx = b \rightarrow Lc = b \rightarrow Ux = c
     $$

### LU分解を用いた連立方程式の解法

1. **LU分解の利用**:
   - 行列 $ A $ を $ LU $ に分解し、$ Ax = b $ を解く際に、まず $ L $ を使って中間変数 $ c $ を求め、次に $ U $ を使って $ x $ を求める手順が説明されています。
   
2. **具体例**:
   - 例として、連立方程式
     $$
     \begin{cases}
     x + y + z = 6 \\
     2x + y - 3z = -7 \\
     2x + y + 4z = 12
     \end{cases}
     $$
     を考えます。
   - LU分解を行い、まず $ L $ を使って
     $$
     \begin{bmatrix}
     1 & 0 & 0 \\
     2 & 1 & 0 \\
     2 & 1 & 1
     \end{bmatrix}
     \begin{bmatrix}
     c_1 \\
     c_2 \\
     c_3
     \end{bmatrix}
     =
     \begin{bmatrix}
     6 \\
     5 \\
     11
     \end{bmatrix}
     $$
     を解いて $ c $ を求め、その後、$ U $ を使って $ x $ を求めています。
   - 最終的な解は $ x = 1, y = 2, z = 3 $ です。
