### 類似行列 (Similar Matrix)

行列 $\mathbf{A}$ と $\mathbf{B}$ が同じ次元の正方行列で、ある可逆行列 $\mathbf{P}$ が存在して $\mathbf{B} = \mathbf{P} \mathbf{A} \mathbf{P}^{-1}$ を満たす場合、$\mathbf{A}$ と $\mathbf{B}$ は類似行列 (similar matrices) であると言います。

#### 類似行列の性質

1. **同一行列は自分自身と類似**  
   $\mathbf{A} = \mathbf{P}^{-1} \mathbf{A} \mathbf{P}$ の場合、$\mathbf{P} = \mathbf{I}$。

2. **類似関係の推移性**  
   $\mathbf{A}$ が $\mathbf{B}$ と類似し、$\mathbf{B}$ が $\mathbf{C}$ と類似するならば、$\mathbf{A}$ は $\mathbf{C}$ と類似する。
$
\mathbf{B} = \mathbf{P}_1 \mathbf{A} \mathbf{P}_1^{-1}, \quad \mathbf{C} = \mathbf{P}_2 \mathbf{B} \mathbf{P}_2^{-1} \implies \mathbf{C} = (\mathbf{P}_2 \mathbf{P}_1) \mathbf{A} (\mathbf{P}_1^{-1} \mathbf{P}_2^{-1}) $

3. **類似行列のランク**  
   類似行列はランクを保存する。
$ \text{rank}(\mathbf{A}) = \text{rank}(\mathbf{P} \mathbf{A} \mathbf{P}^{-1}) = \text{rank}(\mathbf{B}) $

4. **トレース (trace) の保存**  
   類似行列のトレースは同じです。
$ \text{Tr}(\mathbf{P}^{-1} \mathbf{A} \mathbf{P}) = \text{Tr}(\mathbf{A}) $

5. **行列式 (determinant) の保存**  
   類似行列の行列式も等しいです。
$ \text{det}(\mathbf{P}^{-1} \mathbf{A} \mathbf{P}) = \text{det}(\mathbf{A}) $

6. **固有値の等価性**  
   類似行列は同じ固有値を持ちますが、固有ベクトルは異なる場合があります。
$ \mathbf{A} \mathbf{x} = \lambda \mathbf{x} \implies \mathbf{P}^{-1} \mathbf{A} \mathbf{P} (\mathbf{P}^{-1} \mathbf{x}) = \lambda (\mathbf{P}^{-1} \mathbf{x}) $

よって、固有ベクトルは $\mathbf{P}^{-1} \mathbf{x}$ に変換されます。

#### 類似行列の性質

行列 $\mathbf{A}$ と $\mathbf{B}$ が類似行列であるとき、累乗行列 $\mathbf{A}^n$ と $\mathbf{B}^n$ も類似です。
$ \mathbf{B} = \mathbf{P}^{-1} \mathbf{A} \mathbf{P} \implies \mathbf{B}^n = \mathbf{P}^{-1} \mathbf{A}^n \mathbf{P} $
