### 逆行列 (Inverse Matrix)

逆数：

$$
3 \times \frac{1}{3} = 1
$$

行列 $\mathbf{A}$ の逆行列 $\mathbf{A}^{-1}$ とは：

$$
\mathbf{A} \times \mathbf{A}^{-1} = \mathbf{A}^{-1} \times \mathbf{A} = \mathbf{I}
$$

方程式 $\mathbf{A}\mathbf{x} = \mathbf{y}$ の両辺に $\mathbf{A}^{-1}$ をかけると（前から）：

$$
\mathbf{A}^{-1} \mathbf{A} \mathbf{x} = \mathbf{A}^{-1} \mathbf{y}
$$

$$
\mathbf{I} \mathbf{x} = \mathbf{x}
$$

$$
\mathbf{x} = \mathbf{A}^{-1} \mathbf{y}
$$

逆行列が存在しない場合もあることに注意が必要です。

### 対角行列 (Diagonal Matrix)

対角行列の例：

$$
\mathbf{D}  =
\begin{bmatrix}
a_1 & 0 & 0 \\
0 & a_2 & 0 \\
0 & 0 & a_3
\end{bmatrix}
$$
<br>
$$
\mathbf{a} = \text{diag}(\mathbf{D}) =
\begin{bmatrix}
a_1 \\
a_2 \\
a_3
\end{bmatrix}
$$
具体例：

$$
\text{diag}
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{bmatrix}
=
\begin{bmatrix}
1 \\
5 \\
9
\end{bmatrix}
$$

長方形の対角行列（rectangular diagonal matrix）：

$$
\begin{bmatrix}
1 & 0 & 0 & 0 \\
0 & 2 & 0 & 0 \\
0 & 0 & 3 & 0
\end{bmatrix}
$$

### 直交行列 (Orthogonal Matrix)

直交行列の例：

$$
\mathbf{Q} =
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
$$

直交行列の性質：

$$
\mathbf{Q} \mathbf{Q}^T =\mathbf{Q} \mathbf{Q}^{-1} = \mathbf{I}
$$

$$
\mathbf{Q}^T = \mathbf{Q}^{-1}
$$

直交行列において、その行列の転置はその行列の逆行列（transpose）と同じです。

---

### ユニタリ行列 (Unitary Matrix)

ユニタリ行列は、行列がその転置の共役行列である場合に使用されます：

$$
\mathbf{Q}^H \times \mathbf{Q} =\mathbf{Q}^{-1} \times \mathbf{Q} = \mathbf{I}
$$

直交行列では、その行列の転置はその行列のユニタリ行列と同じです。

$$
\mathbf{Q}^H = \mathbf{Q}^{-1}
$$
