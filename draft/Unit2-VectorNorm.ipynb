### ベクトルの内積と正射影

- ベクトル $\mathbf{a}$ を $\mathbf{b}$ に射影したものが $\mathbf{b} \hat{x}$ で、 $\mathbf{b}$ のスカラー$\hat{x}$倍です。
- $\mathbf{a}$ と $\mathbf{b} \hat{x}$ は垂直であることを利用します。これらのベクトルは互いに直交するので、内積は0です。

$$
(\mathbf{a} - \mathbf{b} \hat{x}) \perp \mathbf{b} \hat{x}
$$

$$
\Rightarrow (\mathbf{a} - \mathbf{b} \hat{x})^T \mathbf{b} \hat{x} = 0
$$

ただし $\hat{x} \neq 0$ の場合、

$$
\mathbf{a}^T \mathbf{b} - \hat{x} \mathbf{b}^T \mathbf{b} = 0
$$

$$
\Rightarrow \hat{x} = \frac{\mathbf{a}^T \mathbf{b}}{\mathbf{b}^T \mathbf{b}}
$$

### ノルム (norm)

- **$2$-norm**：$\ell_2$-norm

$$
\|\mathbf{a}\| = \sqrt{\mathbf{a} \cdot \mathbf{a}} = \sqrt{x^2 + y^2 + z^2} = (x^2 + y^2 + z^2)^{\frac{1}{2}}
$$



$$
\mathbf{a} = \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix} \Rightarrow \|\mathbf{a}\|_2 = \sqrt{1^2 + 2^2 + 3^2} = \sqrt{14}
$$

$$
\mathbf{b} - \mathbf{a} = \begin{bmatrix} 2 \\ 3 \end{bmatrix} - \begin{bmatrix} -1 \\ -1 \end{bmatrix}
$$

$$
\|\mathbf{b} - \mathbf{a}\|_2   = \sqrt{(1-2)^2 + (2-1)^2} = \sqrt{(-1)^2 + (1)^2}
$$

- **$1$-norm**:$\ell_1$-norm

$$
\|\mathbf{a}\| = |x| + |y| + |z|
$$



$$
\|\mathbf{a}\|_1 = 1 + 2 + 3 = 6
$$

- **$p$-norm**

$$
\left( |⋅|^p + |⋅|^p  + |⋅|^p  + \ldots \right)^\frac{1}{p}
$$

$$
\|\mathbf{a}\|_p \approx \left( \sum |a_i|^p \right)^{\frac{1}{p}} \quad (p \ge 1)
$$

$$
p \to \infty
$$

$$
\left( |⋅|^\infty + |⋅|^\infty  + |⋅|^\infty  + \ldots \right)^\frac{1}{\infty}=|⋅|
$$
⇒最も大きい値のみが残ります。


- **Infinity norm**

$$
\|\mathbf{a}\|_\infty = \max_i |a_i|
$$

グラフとして以下の条件が成り立つことが示されています。

$$
x + y = 1 \quad y = x + 1
$$

$$
\sqrt{x^2 + y^2} = 1
$$

$$
|x| + |y| = 1
$$

$$
x + y = 1, \quad y = -x + 1
$$

$$
y = x -1
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Figure setup
plt.figure(figsize=(8, 8))
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.axis('equal')  # To maintain aspect ratio

# Plot 1: x^2 + y^2 = 1 (Unit Circle)
theta = np.linspace(0, 2 * np.pi, 100)
x_circle = np.cos(theta)
y_circle = np.sin(theta)
plt.plot(x_circle, y_circle, label=r'$x^2 + y^2 = 1$', color='blue')

# Plot 2: |x| + |y| = 1 (Diamond Shape)
x_diamond = np.linspace(-1, 1, 100)
y_diamond_top = 1 - np.abs(x_diamond)
y_diamond_bottom = np.abs(x_diamond) - 1
plt.plot(x_diamond, y_diamond_top, color='red')
plt.plot(x_diamond, y_diamond_bottom, color='red', label=r'$|x| + |y| = 1$')

# Plot additional lines to form a square
plt.plot([-1, 1], [1, 1], color='green')
plt.plot([-1, 1], [-1, -1], color='green')
plt.plot([1, 1], [1, -1], color='green')
plt.plot([-1, -1], [1, -1], color='green')


# Set labels and legend
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graphs of Given Equations')
plt.legend(loc='best')

# Show the plot
plt.show()
```
