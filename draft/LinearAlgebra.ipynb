**###ベクトル$\mathbf{a}$とベクトル$\mathbf{b}$の内積 (dot product, scalar product, inner product)**

$$\mathbf{a} \cdot \mathbf{b} = \mathbf{a}^T \mathbf{b} = \mathbf{b}^T \mathbf{a}$$

例えば、2つのベクトルが以下のように与えられたら、
$$ \mathbf{a} = \begin{bmatrix} 1 \\ 3 \end{bmatrix}, \quad \mathbf{b} = \begin{bmatrix} 5 \\ 1 \end{bmatrix} $$

その内積は以下のようになります。
$$\mathbf{a} \cdot \mathbf{b} = \mathbf{a}^T \mathbf{b} = \begin{bmatrix} 1 & 3 \end{bmatrix} \begin{bmatrix} 5 \\ 1 \end{bmatrix} = 1 \times 5 + 3 \times 1 = 8$$

ベクトルの内積は次のように表現されることもあります。

$$\mathbf{a} \cdot \mathbf{b} = \|\mathbf{a}\| \|\mathbf{b}\| \cos \theta$$

また、$\mathbf{a} \cdot \mathbf{b}=\|\mathbf{a}\| \cos \theta \|\mathbf{b}\|$と表した場合には、を$\mathbf{b}$ベクトルに**射影（projection）**したときの長さを$\mathbf{b}$ベクトルの長さにかけたものを意味し、逆に、


$\mathbf{a} \cdot \mathbf{b}=\|\mathbf{b}\| \cos \theta \|\mathbf{a}\|$と表現した場合は、$\mathbf{b}$ベクトルを$\mathbf{a}$ベクトルに射影したときの長さを$\mathbf{a}$ベクトルの長さにかけたものになります。


さらに$\mathbf{a}$ベクトル同士の内積は次のようになります。
$$\mathbf{a} \cdot \mathbf{a} = \|\mathbf{a}\|^2 = \|\mathbf{a}\|^2$$

よって、$\mathbf{a}$ベクトルの大きさは次式となります。

$$\|\mathbf{a}\| = \sqrt{\mathbf{a} \cdot \mathbf{a}} = \sqrt{\mathbf{a}^T \mathbf{a}}$$
##単位ベクトル（Unit vector）
大きさが1のベクトル

$$\mathbf{e}_1 =\frac{\mathbf{a}}{\|\mathbf{a}\|}=\frac{\mathbf{a}}{\sqrt{\mathbf{a}^T \mathbf{a}}}$$

また、$\mathbf{e}_2 $と表記し、次のように計算されます。このようにあるベクトルを自分自身の大きさで割ると方向が同じで大きさが1のベクトルが得られます。このベクトルを単位ベクトルといい、この操作をNormalizeといいます。

$$\mathbf{e}_1 = \frac{\mathbf{a}}{\|\mathbf{a}\|}, \mathbf{e}_2 = \frac{\mathbf{b}}{\|\mathbf{b}\|}, \mathbf{e}_3 = \frac{\mathbf{c}}{\|\mathbf{c}\|} $$

##内積の最大・最小

* 2ベクトルの内積が最大 → $\mathbf{a}$と$\mathbf{b}$が同じ向きになったとき→$$ \cos \theta = 1$$
* 2ベクトルの内積が最小 → 直交するとき →$$ \cos \theta = 0$$

$$\mathbf{a} \cdot \mathbf{b} = \|\mathbf{a}\| \|\mathbf{b}\| \cos \theta$$

2つのベクトルの関係性を評価するのに、$ \cos \theta $の値を使うことがあります。この値を**相関係数(Correlation Coefficient)**と呼んだり、**コサイン類似度(Cosine Similarity)**と呼んだりします。


コサイン類似度は、機械学習において頻繁に登場する言葉で、2つのベクトルが「どのくらい似ているか」という類似性を表す尺度です。具体的には、2つのベクトルがなす角のコサイン値のことであり、1なら「似ている」を、-1なら「似ていない」を意味します。主に文書同士の類似性を評価するために使われています。


$$ \cos \theta = \frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{a}\|\|\mathbf{b}\|} $$



* $\mathbf{a}$ベクトルを$\mathbf{b}$ベクトルに射影している部分：$\mathbf{a}$ベクトルが$\mathbf{b}$ベクトルに似ている度合いを表します。

$$\mathbf{a} = \begin{bmatrix} 3 \\ 1 \end{bmatrix}, \quad \mathbf{b} = \begin{bmatrix} 5 \\ 1 \end{bmatrix}$$

$$ \|\mathbf{a}\| \cos \theta =\frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{b}\|}=  \frac{8}{\sqrt{5^2+1^2}} $$



$$\mathbf{a} \cdot \mathbf{b} = \frac{\mathbf{a} \cdot \mathbf{b}}{\|\mathbf{b}\|^2} \mathbf{b} = \frac{8}{\sqrt{52}} \mathbf{b} = \frac{8}{\sqrt{52}} \begin{bmatrix} 5 \\ 1 \end{bmatrix}$$


```python
import numpy as np
import matplotlib.pyplot as plt

# ベクトルを定義
a = np.array([1, 3])
b = np.array([5, 1])

# ベクトルの内積を計算
dot_product = np.dot(a, b)

# ベクトルの長さ（ノルム）を計算
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)

# 単位ベクトルを計算
unit_a = a / norm_a
unit_b = b / norm_b

# aベクトルをbベクトルに投影する
proj_a_on_b = (dot_product / norm_b**2) * b

# グラフを描画
plt.figure(figsize=(8, 8))

# ベクトルaを描画
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='a')

# ベクトルbを描画
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='b')

# ベクトルaをbに投影したものを描画, 破線のパターンを修正
#plt.quiver(0, 0, proj_a_on_b[0], proj_a_on_b[1], angles='xy', scale_units='xy', scale=1, color='g', linestyle='--', dashes=(3, 2), label='Projection of a on b')

# dashes=(3, 2) は、3ポイントの破線と2ポイントの空白を交互に繰り返すパターンを指定します。

# 原点を描画
plt.plot(0, 0, 'ko')

# 軸の範囲を設定
plt.xlim(-1, 6)
plt.ylim(-1, 6)

# グリッドを表示
plt.grid()

# ラベルを表示
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# タイトルを設定
plt.title('Vector Projection')

# グラフを表示
plt.show()
```
