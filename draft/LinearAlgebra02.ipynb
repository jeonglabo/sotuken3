#線形代数（Linear Algebra）

##1次変換行列（線形変換行列）

$ A=\begin{bmatrix}
3 & 1 \\
0 & 2 \\
\end{bmatrix}
 $


線形変換行列$A$によってベクトル$x$を変換してベクトル$y$を得る。


 $ y=Ax$

```python
import numpy as np
import matplotlib.pyplot as plt

# 定義された行列
A = np.array([[3, 1],
              [0, 2]])

# グリッドの範囲
x_range = np.arange(-5, 6)
y_range = np.arange(-2, 9)

# 列ベクトル v1, v2 を取得
v1 = A[:, 0]
v2 = A[:, 1]

# グラフを描画
plt.figure(figsize=(10, 8))

# 基底ベクトル v1 を使ったグリッド
for x in x_range:
    plt.plot(v1[0]*x + v2[0]*y_range, v1[1]*x + v2[1]*y_range, 'r--', linewidth=1)
for y in y_range:
    plt.plot(v1[0]*x_range + v2[0]*y, v1[1]*x_range + v2[1]*y, 'r--', linewidth=1)

# 座標軸
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)

# 列ベクトルを描画
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='red', label='[3 0]')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='blue', label='[1 2]')

# ラベル
plt.xlabel('x')
plt.ylabel('y')

# グラフの範囲
plt.xlim(-4, 10)
plt.ylim(-2, 12)

# 表示
plt.grid(True)
plt.title("Grid in the Coordinate System Defined by Column Vectors of Matrix A")
plt.show()
```

固有値と固有ベクトル

```python
# 固有値と固有ベクトルを求める
eigenvalues, eigenvectors = np.linalg.eig(A)

# 固有ベクトルを描画
for i in range(len(eigenvalues)):
    vec = eigenvectors[:, i] * 5  # 固有ベクトルを5倍して描画
    plt.quiver(0, 0, vec[0].real, vec[1].real, angles='xy', scale_units='xy', scale=1, label=f'Eigenvector {i+1}', color=['green', 'purple'][i])

# グラフの範囲
plt.xlim(-6, 10)
plt.ylim(-2, 12)

# 表示
plt.grid(True)
plt.title("Grid, Column Vectors of Matrix A, and Eigenvectors")
plt.show()

eigenvalues, eigenvectors
```

逆行列

```python
# 行列 A の逆行列を計算する
A_inv = np.linalg.inv(A)
A_inv
```

ランクの計算

```python
# 行列 A のランクを計算する
rank_A = np.linalg.matrix_rank(A)
rank_A
```

行列式の計算

```python
# 行列 A の行列式を計算する
det_A = np.linalg.det(A)
det_A
```

対角化

```python
# 行列 A を対角化する
eigenvalues, eigenvectors = np.linalg.eig(A)

# 対角行列 D
D = np.diag(eigenvalues)

# 行列 P
P = eigenvectors

# 行列 P の逆行列
P_inv = np.linalg.inv(P)

D, P, P_inv
```
