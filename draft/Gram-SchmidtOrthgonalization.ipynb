### 類似行列とGram-Schmidtの直交化


### Gram-Schmidtの直交化法

Gram-Schmidtの直交化法を用いることで、線形独立なベクトル $\mathbf{a}_1, \mathbf{a}_2, \mathbf{a}_3, \mathbf{a}_4$ から直交ベクトルを構成することができます。この方法では、ベクトルのスパンが新たな直交基底を形成します。

#### 手順

1. 最初のベクトル $\mathbf{v}_1$ は、与えられた最初のベクトルそのもの：
$  \mathbf{v}_1 = \mathbf{a}_1 $
   正規化して、$\mathbf{q}_1 = \frac{\mathbf{v}_1}{\|\mathbf{v}_1\|}$ とします。

2. 次のベクトル $\mathbf{v}_2$ は、$\mathbf{a}_2$ から $\mathbf{v}_1$ の成分を除いたもの：
$ \mathbf{v}_2 = \mathbf{a}_2 - \frac{\mathbf{a}_2^T \mathbf{v}_1}{\|\mathbf{v}_1\|^2} \mathbf{v}_1  $

   正規化して、$\mathbf{q}_2 = \frac{\mathbf{v}_2}{\|\mathbf{v}_2\|}$ とします。

3. 三番目のベクトル $\mathbf{v}_3$ は、$\mathbf{a}_3$ から $\mathbf{v}_1$ と $\mathbf{v}_2$ の成分を除いたもの：
$ \mathbf{v}_3 = \mathbf{a}_3 - \mathbf{V}(\mathbf{V}^T \mathbf{V})^{-1} \mathbf{V}^T \mathbf{a}_3 $

   ここで、$\mathbf{V} = [\mathbf{v}_1, \mathbf{v}_2]$ です。これは $\mathbf{a}_3$ を直交基底 $\mathbf{v}_1, \mathbf{v}_2$ に射影し、その射影ベクトルを $\mathbf{a}_3$ から引くことで求めます。

4. 最後のベクトル $\mathbf{v}_4$ も同様に求められます。

---

### Gram-Schmidtの応用

1. 直交基底の構成：
   - ベクトル $\mathbf{v}_1, \mathbf{v}_2, \mathbf{v}_3$ を用いて直交基底を構成します。
   - 新たな基底のスパンが元のベクトルのスパンと一致します。

2. ベクトルの直交基底への射影：
   - 元のベクトル $\mathbf{a}_3$ を直交基底に射影し、その結果を用いて各ベクトルを調整します。

#### 例

ベクトルの射影を用いて、最小二乗問題 $\mathbf{A} \mathbf{x} \approx \mathbf{b}$ の解を求める手法：
$ \hat{\mathbf{x}} = (\mathbf{A}^T \mathbf{A})^{-1} \mathbf{A}^T \mathbf{b} $

この方法により、誤差の二乗和が最小になるような $\mathbf{x}$ を求めます。

---

### まとめ

Gram-Schmidtの直交化法は、ベクトルの直交基底を構築するための基本的な手法であり、最小二乗法や数値解析の分野で重要な役割を果たします。類似行列の性質と組み合わせることで、行列計算の効率化や問題解決の手段としても応用されます。

### Gram-Schmidtの続き

Gram-Schmidtの直交化法を用いて、直交基底を構成する過程は以下の通りです。

#### ベクトル $\mathbf{v}_3$ と $\mathbf{v}_4$ の計算

3番目の直交ベクトル $\mathbf{v}_3$ は、$\mathbf{a}_3$ から $\mathbf{v}_1$ と $\mathbf{v}_2$ の成分を除いたものです。
$ \mathbf{v}_3 = \mathbf{a}_3 - \left( \frac{\mathbf{v}_1 \mathbf{v}_1^T}{\|\mathbf{v}_1\|^2} + \frac{\mathbf{v}_2 \mathbf{v}_2^T}{\|\mathbf{v}_2\|^2} \right) \mathbf{a}_3 $

これを展開すると、
$ \mathbf{v}_3 = \mathbf{a}_3 - \frac{\mathbf{a}_3^T \mathbf{v}_1}{\|\mathbf{v}_1\|^2} \mathbf{v}_1 - \frac{\mathbf{a}_3^T \mathbf{v}_2}{\|\mathbf{v}_2\|^2} \mathbf{v}_2 $

4番目のベクトル $\mathbf{v}_4$ は、$\mathbf{a}_4$ から他のベクトルの成分を除いたものです。
$ \mathbf{v}_4 = \mathbf{a}_4 - \frac{\mathbf{a}_4^T \mathbf{v}_1}{\|\mathbf{v}_1\|^2} \mathbf{v}_1 - \frac{\mathbf{a}_4^T \mathbf{v}_2}{\|\mathbf{v}_2\|^2} \mathbf{v}_2 - \frac{\mathbf{a}_4^T \mathbf{v}_3}{\|\mathbf{v}_3\|^2} \mathbf{v}_3 $

正規化して直交基底の4つ目のベクトル $\mathbf{q}_4$ を得ます。
$ \mathbf{q}_4 = \frac{\mathbf{v}_4}{\|\mathbf{v}_4\|} $
